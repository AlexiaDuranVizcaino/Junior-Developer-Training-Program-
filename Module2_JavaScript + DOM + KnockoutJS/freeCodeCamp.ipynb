## JS Basics — Cheat Sheet

- **Comments:** `// inline` · `/* multi-line */`
- **8 data types:** undefined, null, boolean, string, symbol, bigint, number, object
- **Declare:** `var name;` (ends with `;`; no spaces, no leading number)
- **Assign `=`:** right side runs first → stored on left
- **Init:** `var x = 9;` (declare + assign in one line)
- **String:** chars in `" "` or `' '`
- **Uninitialized** → `undefined`; math on it → `NaN`; with string → `"undefined"`

```python
var a = 7;            // declare + init
var b = a;            // copy value -> 7
var name = "Alexia";  // string

var c = 5;
c = c + 1;            // 6  (init avoids undefined/NaN)
```
