# 🚀 Microsoft Learn: Automate with GitHub Actions

## 📚 Module Introduction

- **GitHub Actions**: Automate code delivery (build, test, deploy) on GitHub.
- **Scenario**: Team needs fast, high-quality web deployment; automate infrastructure with CI/CD and IaC.
- **Why GitHub Actions?** Already using GitHub, want integrated automation.

### 🎯 Learning Objectives

- Understand GitHub Actions and action types.
- Identify workflow file components.
- Plan automation for the software lifecycle.
- Create and run a container action triggered by a push event.


## 🔧 How GitHub Actions Automate Development Tasks

### 📉 GitHub Decreases Time: Idea → Deployment

**Two main features:**
- **Communication**: Pull requests, code reviews, issues, project boards, wikis, notifications
- **Automation**: GitHub Actions for CI/CD workflows at every step

### 🔄 Typical Automated Tasks

- **Unit tests**: Ensure code passes all tests
- **Code quality checks**: Meet organization standards
- **Security scans**: Check dependencies for vulnerabilities
- **Build**: Integrate new source code from multiple contributors
- **Integration tests**: Verify software passes integration tests
- **Versioning**: Specify build version
- **Delivery**: Move binaries to appropriate filesystem location
- **Deploy**: Deploy binaries to servers
- **Error reporting**: Report issues to proper individual/team

### 🎯 What is GitHub Actions?

**Packaged scripts** to automate tasks in software-development workflows on GitHub.

- **Trigger options**: Push, pull request, scheduled intervals, or manual
- **Location**: `.github/workflows/` directory
- **Format**: YAML files (`.yml`)
- **Result**: Reliable, sustainable automated workflows → significant decrease in development time

### 📦 Where to Find Actions

- **Actions tab**: Quick setup with templates (Configure button)
- **GitHub Marketplace**: Discover and purchase tools
- **Open-source repos**: GitHub Actions organization and other public repos
- **Write your own**: Custom actions (can publish to Marketplace)

### ✅ Using Open-Source Actions - Best Practices

- Review `action.yml` file for inputs, outputs, and code verification
- Check if action is in GitHub Marketplace
- Check if action is **verified** by GitHub (but still review it)
- Specify version using Git ref, SHA, or tag

### 🧩 Types of GitHub Actions

**Container Actions:**
- Environment is part of the action's code
- Runs on Linux only (GitHub-hosted)
- Supports many different languages

**JavaScript Actions:**
- Environment specified separately
- Runs on VM in cloud or on-premises
- Supports Linux, macOS, Windows

**Composite Actions:**
- Combine multiple workflow steps within one action
- Bundle multiple run commands
- Execute bundled commands as single step

### 📝 Anatomy of a GitHub Action

**Simple action example:**
```yaml
steps:
  - uses: actions/checkout@v1
  - name: npm install and build webpack
    run: |
      npm install
      npm run build
```

**Container action example:**
```yaml
name: "Hello Actions"
description: "Greet someone"
author: "octocat@github.com"

inputs:
  MY_NAME:
    description: "Who to greet"
    required: true
    default: "World"

runs:
  uses: "docker"
  image: "Dockerfile"

branding:
  icon: "mic"
  color: "purple"
```

**Key sections:**
- **`inputs`**: Variables/parameters set in workflow
- **`runs`**: Specifies docker and path to Docker image file
- **`branding`**: Personalizes action in GitHub Marketplace

## 🔄 GitHub Actions Workflows

### 📋 What is a Workflow?

A process to automate software-development lifecycle tasks in your repository.

- Build, test, package, release, deploy projects
- Add actions to `.yml` file in `.github/workflows/` directory

### 🏗️ Workflow Structure Example

```yaml
name: A workflow for my Hello World file
on: push
jobs:
  build:
    name: Hello world action
    runs-on: ubuntu-latest
    steps:
    - uses: actions/checkout@v1
    - uses: ./action-a
      with:
        MY_NAME: "Mona"
```

### 🎬 Workflow Components

- **`name`**: Workflow name
- **`on`**: Trigger event(s) that start the workflow
- **`jobs`**: Must have at least one job
- **`runs-on`**: Specifies the runner (GitHub-hosted or self-hosted)
- **`steps`**: Tasks to complete in each job
- **`uses`**: References an action (e.g., `actions/checkout@v1` or `./action-a`)
- **`with`**: Sets input variables (e.g., `MY_NAME: "Mona"`)

### ⚡ Trigger Events (`on`)

**Single event:**
```yaml
on: push
```

**Multiple events:**
```yaml
on: [push, pull_request]
```

**Event configuration map:**
```yaml
on:
  push:
    branches:
      - main
  pull_request:
    branches:
      - main
  page_build:
  release:
    types:
      - created
```

**Key points:**
- Event triggers on all activity types unless specified
- Can schedule workflows or restrict to specific files, tags, branches

### 🔗 Referencing Actions in Workflows

**Docker Hub image:**
```yaml
steps:
  - name: Run a Docker action
    uses: docker://<docker-image-name>:<tag>
```

**Public repository:**
```yaml
steps:
  - name: Use a public action
    uses: actions/checkout@v3
```

**Same repository (local action):**
```yaml
steps:
  - name: Use a local action
    uses: ./path-to-action
```

**Enterprise marketplace:**
```yaml
steps:
  - name: Use an enterprise marketplace action
    uses: enterprise-org/action-name@v1
```

⚠️ **Security Best Practice:** Use full commit SHA instead of tags
- Example: `uses: actions/checkout@c2c1744e079e0dd11c8e0af4a96064ca4f6a2e9e`
- Ensures workflow always uses exact same code
- Always specify version (Git ref, SHA, or tag) for consistency

## 🖥️ Runners

### What is a Runner?

A server with GitHub Actions runner application installed.

### GitHub-Hosted vs Self-Hosted Runners

**GitHub-Hosted Runners:**
- Each job runs in fresh instance of virtual environment
- Syntax: `runs-on: ubuntu-latest`
- Quick and simple
- Limited options

**Self-Hosted Runners:**
- Custom local environment (on-premises or cloud)
- Syntax: `runs-on: [self-hosted, linux, ARM32]`
- Highly configurable
- Custom hardware configuration (more CPU/memory)
- Install software available on local network
- Choose OS not offered by GitHub

**Example:**
```yaml
# GitHub-hosted
runs-on: ubuntu-latest

# Self-hosted (Linux + ARM32)
runs-on: [self-hosted, linux, ARM32]
```

### 🚀 Larger Runners (GitHub-Hosted)

**Purpose:** Handle resource-intensive workflows efficiently

**Ideal for:**
- Compiling large codebases
- Comprehensive test suites (integration, end-to-end)
- Processing large datasets (data analysis, ML)
- Complex dependencies or large binary outputs
- High-performance simulations/computational modeling
- Video encoding, rendering, multimedia processing

**Usage example:**
```yaml
jobs:
  build:
    runs-on: ubuntu-latest-16core    # 16 vCPUs, 64GB RAM
    steps:
      - uses: actions/checkout@v2
      - name: Build project
        run: make build
```

**Benefits:**
- Increased CPU, memory, disk space
- Compatible with existing workflows
- Same preinstalled tools as standard runners

### 📊 Managing Larger Runners

**Monitoring Usage:**
- View GitHub Actions usage page in repository/organization settings
- Insights: number of jobs, total runtime, associated costs

**Managing Access:**
- Configure repository/organization-level policies
- Ensure only authorized workflows/teams use high-resource runners

**Cost Management:**
- Use larger runners only when needed
- Optimize workflows to reduce runtime
- Monitor billing details regularly

**Scaling Strategies:**
- Use self-hosted runners for predictable workloads
- Split workflows into smaller jobs across standard runners

### ⚠️ Usage Limits

GitHub Actions has usage limits based on:
- GitHub plan (Free, Pro, Enterprise)
- Runner type (GitHub-hosted vs self-hosted)

💡 Check GitHub documentation for specific limits, billing, and administration details.

## 🧩 Components of GitHub Actions

### 📊 Component Hierarchy

**Event** → **Workflow** → **Job** → **Steps** → **Actions**

An event triggers a workflow, which contains jobs that use steps to dictate which actions run.

---

### 🔄 Workflows

**Definition:** Automated process added to your repository

**Characteristics:**
- Must have at least one job
- Can be triggered by different events
- Located in `.github/workflows/` directory

**Use cases:**
- Build, test, package, release, or deploy projects

---

### 💼 Jobs

**Definition:** Major section of the workflow associated with a runner

**Key points:**
- First major component within workflow
- Runs on a machine or in a container
- Specified with `runs-on:` attribute

**Example:**
```yaml
jobs:
  build:
    runs-on: ubuntu-latest
    steps:
      # steps go here
```

**Runner types:**
- GitHub-hosted (e.g., `ubuntu-latest`)
- Self-hosted (e.g., `[self-hosted, linux, ARM32]`)

---

### 👣 Steps

**Definition:** Individual task that can run commands in a job

**Characteristics:**
- Part of a job
- Can use actions or run commands
- Execute sequentially

**Example:**
```yaml
steps:
  - uses: actions/checkout@v2
  - uses: ./action-a
  - run: npm install -g bats
```

---

### ⚡ Actions

**Definition:** Standalone commands executed inside your workflow

**Types of actions:**
- **GitHub actions**: Official actions (e.g., `actions/checkout@v2`)
- **Community actions**: Third-party actions from GitHub Marketplace
- **Custom actions**: Your own actions (e.g., `./action-a`)

**Usage options:**
- Reference actions with `uses:`
- Run shell commands with `run:`

**Examples:**
```yaml
# Using GitHub action
- uses: actions/checkout@v2

# Using custom action
- uses: ./action-a

# Running command
- run: npm install -g bats
```

---

### 🔗 How Components Work Together

```yaml
# Event triggers the workflow
on: push

# Workflow contains job(s)
jobs:
  build:                        # Job name
    runs-on: ubuntu-latest      # Runner
    steps:                      # Steps in the job
      - uses: actions/checkout@v2     # Action
      - run: npm install              # Command action
      - uses: ./custom-action         # Custom action
```

**Flow:**
1. **Event occurs** (e.g., push to repository)
2. **Workflow is triggered**
3. **Job starts** on specified runner
4. **Steps execute** sequentially
5. **Actions run** within each step

## ⚙️ Configure a GitHub Actions Workflow

### 📅 Scheduled Events

**Trigger workflows at specific UTC times using POSIX cron syntax**

**Cron syntax format:** Five fields representing units of time
```
* * * * *
│ │ │ │ │
│ │ │ │ └─ Day of week (0-7, SUN-SAT)
│ │ │ └─── Month (1-12)
│ │ └───── Day of month (1-31)
│ └─────── Hour (0-23)
└───────── Minute (0-59)
```

**Examples:**

Run every 15 minutes:
```yaml
on:
  schedule:
    - cron: '*/15 * * * *'
```

Run every Sunday at 3:00 AM:
```yaml
on:
  schedule:
    - cron: '0 3 * * SUN'
```

**Key points:**
- Shortest interval: Every 5 minutes
- Runs on latest commit of default/base branch
- Can use operators for ranges and values

---

### 🖱️ Manual Events

#### workflow_dispatch

**Trigger workflows manually via GitHub UI or REST API**

**Features:**
- Choose which branch to run on
- Set optional inputs displayed as form elements in UI

**Example:**
```yaml
on:
  workflow_dispatch:
    inputs:
      logLevel:
        description: 'Log level'
        required: true
        default: 'warning'
      tags:
        description: 'Test scenario tags'
```

#### repository_dispatch

**Trigger workflows from external systems via GitHub API**

**Purpose:** Allows triggering workflows for activity outside of GitHub

**Requirements:**
1. Send POST request to GitHub endpoint
2. Configure workflow to use `repository_dispatch` event

**API request example:**
```bash
curl -X POST \
  -H "Accept: application/vnd.github.v3+json" \
  https://api.github.com/repos/octocat/hello-world/dispatches \
  -d '{"event_type":"event_type"}'
```

**Workflow configuration:**
```yaml
on:
  repository_dispatch:
    types: [opened, deleted]
```

---

### 🔔 Webhook Events

**Trigger workflows when specific webhook events occur on GitHub**

**Features:**
- Most webhook events have multiple activities
- Specify activity types to trigger workflow

**Example:**
```yaml
on:
  check_run:
    types: [rerequested, requested_action]
```

### 🎯 repository_dispatch Deep Dive

**Custom event for external triggers**

**Use Cases:**
- Trigger workflows from external CI/CD tools
- Coordinate multi-repo deployments (Repo A → triggers Repo B)
- Start automation from external events (webhooks, alerts, CRON jobs)
- Chain workflow executions between repositories

**Workflow Example:**
```yaml
name: Custom Dispatch Listener

on:
  repository_dispatch:
    types: [run-tests, deploy-to-prod]

jobs:
  run:
    runs-on: ubuntu-latest
    steps:
      - name: Echo the payload
        run: |
          echo "Event type: ${{ github.event.action }}"
          echo "Payload value: ${{ github.event.client_payload.env }}"
```

**Key elements:**
- **`types`**: Optional custom event types (e.g., `run-tests`, `deploy-to-prod`)
- **`github.event.client_payload`**: Access custom data passed in dispatch event
- **`github.event.action`**: Name of the event_type sent

**API Endpoint:**
```
POST https://api.github.com/repos/OWNER/REPO/dispatches
```

**Authorization:**
- Requires Personal Access Token (PAT) with `repo` scope
- For organizations, ensure proper access settings

**Sample API call:**
```bash
curl -X POST \
  -H "Accept: application/vnd.github+json" \
  -H "Authorization: token YOUR_GITHUB_TOKEN" \
  https://api.github.com/repos/OWNER/REPO/dispatches \
  -d '{"event_type":"run-tests","client_payload":{"env":"staging"}}'
```

**Payload Structure:**
```json
{
  "event_type": "run-tests",
  "client_payload": {
    "env": "staging"
  }
}
```

#### Parameters Breakdown

**`event_type`** (Required):
- **Type:** String
- **Purpose:** Custom name for the event, maps to `types` in workflow trigger
- **Examples:** `"deploy"`, `"run-tests"`, `"sync-db"`, `"build-docker"`
- **Access in workflow:** `${{ github.event.action }}`

**`client_payload`** (Optional):
- **Type:** Object (free-form JSON)
- **Purpose:** Send custom data to workflow
- **Use cases:** Multi-environment deployments, versioned releases, passing context
- **Access in workflow:** `${{ github.event.client_payload.key }}`

**Example with multiple fields:**
```json
{
  "event_type": "deploy-to-prod",
  "client_payload": {
    "env": "production",
    "build_id": "build-456",
    "initiator": "admin_user",
    "services": ["web", "api", "worker"]
  }
}
```

**Accessing in workflow:**
```yaml
- name: Show payload values
  run: |
    echo "Environment: ${{ github.event.client_payload.env }}"
    echo "Version: ${{ github.event.client_payload.build_id }}"
    echo "Initiator: ${{ github.event.client_payload.initiator }}"
```

### 🔀 Use Conditional Keywords

**Evaluate expressions to determine if step should run**

**Syntax:** `${{ <expression> }}`

**Example:**
```yaml
name: CI
on: push
jobs:
  prod-check:
    if: github.ref == 'refs/heads/main'
    runs-on: ubuntu-latest
    steps:
      # steps only run if on main branch
```

**Key points:**
- Access context information and evaluate expressions
- Commonly used with `if` keyword
- Some expressions (like `if`) can omit `${{ }}` syntax
- GitHub automatically evaluates common expressions

### 🛑 Disable and Delete Workflows

**Temporarily disable workflows without deleting files**

**Methods:**
- Via GitHub UI (Actions tab)
- Via GitHub REST API

**When to disable:**
- Workflow error producing too many/wrong requests
- Workflow consuming too many minutes
- Workflow sending requests to a down service
- Working on fork with unnecessary workflows

**Cancel workflow in progress:**
- GitHub UI: Actions tab
- API: `DELETE /repos/{owner}/{repo}/actions/runs/{run_id}`

⚠️ Canceling a workflow cancels all jobs and steps within that run

### 🔐 Use Specific Versions of Actions

**Best practice: Reference specific versions instead of just action name**

**Why?**
- Safeguard against unexpected changes
- Prevent workflow breakage from action updates
- Ensure stability and security

**Referencing methods:**

```yaml
steps:
  # Specific commit (Most secure)
  - uses: actions/setup-node@c46424eee26de4078d34105d3de3cc4992202b1e
  
  # Major version
  - uses: actions/setup-node@v1
  
  # Minor version
  - uses: actions/setup-node@v1.2
  
  # Branch (Less safe - gets latest changes)
  - uses: actions/setup-node@main
```

**Security recommendations:**
- 🔒 **Most secure:** Commit SHA of released action
- ⚠️ **Less secure:** Branch reference (gets latest changes)
- ✅ **Recommended:** Version number or commit SHA for stability

💡 Using commit SHA ensures you run exact same code every time
