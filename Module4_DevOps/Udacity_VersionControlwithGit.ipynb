# 🚀 Version control with Git

Version control is software that helps you record, manage, and recover changes to files over time. It creates a detailed history of a project (commits), enables collaboration through branching and merging, and lets you restore previous states when needed. This course uses Git, a distributed version control system.

## 📚 Git and Version Control Terminology

The key terms are:

- **🔄 Version Control System / Source Code Manager (VCS / SCM):** A tool that manages different versions of source code. Git is an SCM/VCS (https://git-scm.com/).

- **📸 Commit:** A snapshot of your project's files at a point in time. Commits are the fundamental unit in Git (like save points).

- **📦 Repository / repo:** A directory containing your project and the Git metadata. Repositories can be local or remote and are composed of commits.

- **📁 Working Directory:** The files you see and edit in your filesystem — distinct from the saved commits in the repository.

- **🔀 Checkout:** The action of copying content from the repository to the Working Directory (switching what files you have checked out).

- **📋 Staging Area / Index:** A place where changes are prepared before being included in the next commit. Think of it as a prep table for the next snapshot.

- **🔑 SHA:** A 40-character hash (hex string) that uniquely identifies a commit, calculated from its contents (e.g., e2adf8ae3e...).

- **🌳 Branch:** A separate line of development that diverges from the main line. Branches let you make commits independently and switch between different work streams.

**⚙️ Typical Workflow (high-level):** edit files in the Working Directory → stage changes in the Index → commit to create snapshots in the repository → use branches to manage parallel development and checkout to switch contexts.

## 📁 Create A Repo From Scratch

### 🎯 Quick Overview
Initialize a Git repository to start version control on your project.

### 📝 Essential Terminal Commands
| Command | Purpose |
|---------|---------|
| `ls` | 📋 List files and directories |
| `mkdir` | 📂 Create a new directory |
| `cd` | 🚀 Change directories |
| `rm` | 🗑️ Remove files/directories |
| `pwd` | 📍 Print working directory |

### 🔧 Step-by-Step Setup

#### 1️⃣ **Create Project Structure**
```bash
mkdir -p udacity-git-course/new-git-project
cd udacity-git-course/new-git-project
```

#### 2️⃣ **Initialize Repository**
```bash
git init
```
✅ Creates a hidden `.git` directory (the heart of your repo!)

### 📦 What's Inside `.git` Directory?

| File/Directory | 🔍 Purpose |
|---|---|
| **config** | 🛠️ Project-specific Git settings (e.g., user email per repo) |
| **objects** | 💾 Stores all commits and project snapshots |
| **refs** | 🏷️ Pointers to commits (branches & tags) |
| **hooks** | 🪝 Scripts for Git workflow events (advanced) |
| **description** | 📄 GitWeb program use only (can ignore) |
| **info** | ℹ️ Global excludes file |

⚠️ **DO NOT EDIT FILES IN .git!** This could break your repository!

### 🎓 Key Takeaway
`git init` → Creates empty repository → Sets up hidden `.git` directory → Ready for commits!

## 🔄 Why Clone? 

### 🎯 What is Cloning?

**Clone** = Make an identical copy of an existing Git repository

### 💡 The Problem & Solution

**Problem:** 🔁 Every new project requires the same repetitive steps:
- Create folder structure (js, css, img)
- Create boilerplate files
- Setup configuration files
- Configure linter & code editor
- ...repeat for EVERY project = 😫 Time-wasting!

**Solution:** 📋 Create a **starter template repository once**, then clone it for every new project!

### 📥 Git Clone Command

```bash
git clone <path-to-repository>
```

**Example:**
```bash
git clone https://github.com/udacity/course-git-blog-project
```

#### Optional: Clone With Custom Name

```bash
git clone <path-to-repository> <custom-directory-name>
```

**Example:**
```bash
git clone https://github.com/udacity/course-git-blog-project blog-project
```

### 🔍 What Happens When You Clone?

| Step | What Git Does |
|------|---|
| 1️⃣ | Creates new directory (repo name or custom name) |
| 2️⃣ | Downloads all files & commit history |
| 3️⃣ | Counts remote objects & compresses them |
| 4️⃣ | Unpacks repository in new directory |
| 5️⃣ | **⚠️ Stays in current directory (doesn't auto-enter repo)** |

### ⚠️ CRITICAL: Nested Repositories Warning 🚫

```
❌ DON'T DO THIS:
$ pwd
/path/to/git-repo        ← Already in a Git repo!
$ git clone <url>        ← Creates nested repo = DISASTER!

✅ DO THIS INSTEAD:
$ cd ~                   ← Exit any Git repo first
$ git clone <url>        ← Clone safely here
```

### 📍 After Cloning - Enter the Repo!

```bash
$ git clone <url>
$ cd course-git-blog-project    ← DON'T FORGET THIS!
$ pwd                           ← Verify you're in repo
```

### 🎓 Quick Summary Table

| Feature | Details |
|---------|---------|
| **Command** | `git clone <url>` |
| **Creates** | New directory + repository copy |
| **Copies** | All files + complete commit history |
| **Default Name** | Same as source repo |
| **Custom Name** | Pass 2nd argument to clone |
| **Location** | Current working directory |
| **Auto-enter?** | ❌ NO - Must `cd` manually! |

### ✅ Git Clone Checklist

- [ ] Verify you're NOT already in a Git repo (`git status` should error)
- [ ] Run `git clone <url>` or `git clone <url> <name>`
- [ ] Run `cd <directory-name>` to enter repo
- [ ] Verify with `git status` (should show branch info)
- [ ] Ready to work! 🚀

## 🔍 Git Status - Read Git's Mind!

### 🎯 What is `git status`?

The **key to understanding Git!** It tells you what Git is thinking and the state of your repository at any moment.

### 💡 Why Use It?

- Know when Git is ready for commands
- Check if commands worked
- Avoid making incorrect assumptions
- **Use it AFTER EVERY command** 🔄

### 📋 The Command

```bash
git status
```

### 🏠 Resting State (Cloned Repository)

When you clone a repo with commits:

```
On branch main
Your branch is up-to-date with 'origin/main'.
nothing to commit, working directory clean
```

**Meaning:**
- ✅ You're on `main` branch
- ✅ Synced with remote repo
- ✅ No changes pending
- ✅ Everything is stable

### 🆕 New Repository State (Empty)

When you initialize a new repo (`git init`):

```
On branch main

No commits yet

nothing to commit (create/copy files and use "git add" to track)
```

**Meaning:**
- 📍 You're on `main` branch (Git 2.28+) or `master` (older versions)
- ⚠️ **No commits exist yet** (this is normal!)
- 📁 Ready to add files

### 📊 What `git status` Tells You

| Info | Shows |
|------|-------|
| **Current Branch** | Which branch you're on |
| **Commit Status** | Number of commits or "No commits yet" |
| **Working Directory** | New files, modified files, deletions |
| **Staging Area** | Files ready to be committed |
| **Sync Status** | If you're ahead/behind remote repo |

### 🧪 Check for Commits

Want to see all commits?

```bash
git log
```

**In empty repo:** `fatal: your current branch 'main' does not have any commits yet` (just means no commits - not an error!)

### ✅ The Golden Rule

```
After EVERY git command → run git status
```

This helps you:
- Learn how Git works 🧠
- Avoid mistakes 🛡️
- Understand your repo state 📍

### 🎓 Git Status Checklist

- [ ] Run `git status` after cloning
- [ ] Run `git status` after modifying files
- [ ] Run `git status` before committing
- [ ] Check branch name in output
- [ ] Verify "working directory clean" when done
- [ ] Get into the habit! 🔄
