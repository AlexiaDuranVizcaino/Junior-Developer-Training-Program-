# 🚀 Version control with Git

Version control is software that helps you record, manage, and recover changes to files over time. It creates a detailed history of a project (commits), enables collaboration through branching and merging, and lets you restore previous states when needed. This course uses Git, a distributed version control system.

## 📚 Git and Version Control Terminology

The key terms are:

- **🔄 Version Control System / Source Code Manager (VCS / SCM):** A tool that manages different versions of source code. Git is an SCM/VCS (https://git-scm.com/).

- **📸 Commit:** A snapshot of your project's files at a point in time. Commits are the fundamental unit in Git (like save points).

- **📦 Repository / repo:** A directory containing your project and the Git metadata. Repositories can be local or remote and are composed of commits.

- **📁 Working Directory:** The files you see and edit in your filesystem — distinct from the saved commits in the repository.

- **🔀 Checkout:** The action of copying content from the repository to the Working Directory (switching what files you have checked out).

- **📋 Staging Area / Index:** A place where changes are prepared before being included in the next commit. Think of it as a prep table for the next snapshot.

- **🔑 SHA:** A 40-character hash (hex string) that uniquely identifies a commit, calculated from its contents (e.g., e2adf8ae3e...).

- **🌳 Branch:** A separate line of development that diverges from the main line. Branches let you make commits independently and switch between different work streams.

**⚙️ Typical Workflow (high-level):** edit files in the Working Directory → stage changes in the Index → commit to create snapshots in the repository → use branches to manage parallel development and checkout to switch contexts.

## 📁 Create A Repo From Scratch

### 🎯 Quick Overview
Initialize a Git repository to start version control on your project.

### 📝 Essential Terminal Commands
| Command | Purpose |
|---------|---------|
| `ls` | 📋 List files and directories |
| `mkdir` | 📂 Create a new directory |
| `cd` | 🚀 Change directories |
| `rm` | 🗑️ Remove files/directories |
| `pwd` | 📍 Print working directory |

### 🔧 Step-by-Step Setup

#### 1️⃣ **Create Project Structure**
```bash
mkdir -p udacity-git-course/new-git-project
cd udacity-git-course/new-git-project
```

#### 2️⃣ **Initialize Repository**
```bash
git init
```
✅ Creates a hidden `.git` directory (the heart of your repo!)

### 📦 What's Inside `.git` Directory?

| File/Directory | 🔍 Purpose |
|---|---|
| **config** | 🛠️ Project-specific Git settings (e.g., user email per repo) |
| **objects** | 💾 Stores all commits and project snapshots |
| **refs** | 🏷️ Pointers to commits (branches & tags) |
| **hooks** | 🪝 Scripts for Git workflow events (advanced) |
| **description** | 📄 GitWeb program use only (can ignore) |
| **info** | ℹ️ Global excludes file |

⚠️ **DO NOT EDIT FILES IN .git!** This could break your repository!

### 🎓 Key Takeaway
`git init` → Creates empty repository → Sets up hidden `.git` directory → Ready for commits!

## 🔄 Git Clone

### 🎯 What is Cloning?

**Clone** = Make an identical copy of an existing Git repository (files + commit history)

💡 **Why Clone?** Reuse template projects instead of recreating folder structures and configs for every new project.

### 📥 Git Clone Command

```bash
git clone <path-to-repository>
git clone <path-to-repository> <custom-directory-name>  # Optional: custom name
```

**Example:**
```bash
git clone https://github.com/udacity/course-git-blog-project
git clone https://github.com/udacity/course-git-blog-project blog-project
```

### ⚠️ Important Notes

- **Nested repos:** Don't clone inside an existing Git repo! Navigate to a safe directory first.
- **After cloning:** Use `cd <directory-name>` to enter the repo (Git doesn't auto-enter).
- **Verification:** Run `git status` to confirm you're in the repo.

### 📋 Quick Reference

| Command | Result |
|---------|--------|
| `git clone <url>` | Creates directory + copies all files & history |
| `git clone <url> <name>` | Clones with custom directory name |

## 🔍 Git Status

### 🎯 What is `git status`?

Shows the current state of your repository: branch, changes, staging area, and sync status.

💡 **Golden Rule:** Run `git status` after EVERY command to understand what Git is doing.

### 📋 The Command

```bash
git status
```

### 📊 What It Shows

| Info | Details |
|------|---------|
| **Current Branch** | Which branch you're on |
| **Commit Status** | "No commits yet" or number of commits |
| **Working Directory** | New/modified/deleted files |
| **Staging Area** | Files ready to commit |
| **Sync Status** | Ahead/behind remote repo |

### 🏠 Common Outputs

**Clean repo (no changes):**
```
On branch main
nothing to commit, working directory clean
```

**New repo (no commits yet):**
```
On branch main
No commits yet
nothing to commit (create/copy files and use "git add" to track)
```

## 📜 Review a Repo's History

### 🎯 The `git log` Command

Displays all commits in a repository with their details.

```bash
git log
```

### 📊 What It Shows (By Default)

| Info | Description |
|------|-------------|
| **SHA** | Unique commit identifier (hash) |
| **Author** | Who made the commit |
| **Date** | When the commit was made |
| **Message** | Commit description |

### 🧭 Navigating Git Log (Less Pager)

Git log uses the **Less** pager to display information:

| Key | Action |
|-----|--------|
| `j` or `↓` | Move down one line |
| `k` or `↑` | Move up one line |
| `d` | Move down half page |
| `u` | Move up half page |
| `f` or `Space` or `Page Down` | Move down full page |
| `b` or `Page Up` | Move up full page |
| `q` | **Quit** (return to prompt) |

### 💡 Key Points

- **First action in a repo:** Always run `git status` to see the current state
- **Git log shows history:** All commits with complete information
- **SHA identification:** You can use first 7 characters of SHA to identify commits
- **Tracking changes:** Find who made changes, when, and why through commit messages

### 📝 Example Output

```
commit f9720a9d4e1e1e1e1e1e1e1e1e1e1e1e1e1e1e1e
Author: John Doe <john@example.com>
Date:   Mon Jan 15 14:30:22 2024 -0500

    Add expedition heading to homepage
    
commit 8aa6668e2e2e2e2e2e2e2e2e2e2e2e2e2e2e2e2e
Author: Jane Smith <jane@example.com>
Date:   Sun Jan 14 10:15:30 2024 -0500

    Convert social links from text to images
```

### 🔍 `git log -p` (--patch)

Shows **exact changes** made to files in each commit (the actual code differences).

```bash
git log -p
# or
git log --patch
```

**What it displays:**
- Files that have been modified
- Location of lines added/removed
- **Actual changes** made (line by line)
  - Lines with `-` (red) = removed
  - Lines with `+` (green) = added

**Example Output Explanation:**

```diff
commit a3dc99a Center content on page
Author: Richard Kalehoff <richardkalehoff@gmail.com>
Date:   Wed Dec 7 13:24:15 2016 -0500

    Center content on page

diff --git a/css/app.css b/css/app.css
index 12345..67890 100644
--- a/css/app.css
+++ b/css/app.css
@@ -15,83 +15,85 @@
 .container {
-  width: 100%;
+  width: 95%;
+  max-width: 1200px;
+  margin: 0 auto;
 }
```

**Key Output Elements:**
- 🔵 **File name**: `css/app.css`
- ❤️ **Version markers**: `---` (old) and `+++` (new)
- 🔍 **Line info**: `@@ -15,83 +15,85 @@`
  - Old: started at line 15, had 83 lines
  - New: starts at line 15, has 85 lines
- ✏️ **Changes**: red `-` (removed), green `+` (added)

### 🔧 Useful Flags

| Flag | Description | Usage |
|------|-------------|-------|
| `-p --stat` | Combines patch + statistics | See changes AND file stats |
| `-w` | Ignores whitespace changes | Useful when code is re-indented |
| `-p -2` | Shows only last 2 commits | Limit output |

**Example with `-w` flag:**
```bash
git log -p -w
```
**Why `-w`?** When code is re-indented, all lines show as removed/added. The `-w` flag ignores whitespace/indentation changes, showing only real code changes.

**Use for:** Understanding exactly what code was modified, added, or removed in each commit

### 🎯 Viewing A Specific Commit

Two ways to view a specific commit without scrolling through all history:

#### 1️⃣ `git log` with SHA

```bash
git log -p fdf5493
```

Shows the specified commit **and all commits before it**.

#### 2️⃣ `git show` (Recommended)

```bash
# Shows most recent commit
git show

# Shows specific commit
git show fdf5493
```

Shows **only one commit** (default = most recent).

**Output includes:**
- Commit SHA
- Author
- Date
- Commit message
- Patch information (actual changes)

**Useful flags:**
```bash
git show --stat fdf5493      # Show file statistics
git show -p fdf5493          # Show patch (default)
git show -w fdf5493          # Ignore whitespace
git show --stat -p fdf5493   # Both statistics and patch
```

## 📝 Add Commits to a Repo

This section covers the essential commands for adding files and creating commits in a Git repository.

### 📦 Git Add

#### Git Add Recap

The `git add` command is used to move files from the Working Directory to the Staging Index.

```bash
$ git add <file1> <file2> … <fileN>
```

**This command:**
- Takes a space-separated list of file names
- Alternatively, the period `.` can be used in place of a list of files to tell Git to add the current directory (and all nested files)

**Examples:**
```bash
# Add specific files
$ git add index.html css/app.css js/app.js

# Add all files in current directory
$ git add .
```

**💡 Important Notes:**
- Running `git add` produces no output when successful
- Use `git status` to verify files have been staged
- Files in the "Changes to be committed" section are in the Staging Index
- Use `git rm --cached <file>` to unstage a file if you added it by mistake

### 💾 Git Commit

#### Git Commit Recap

The `git commit` command takes files from the Staging Index and saves them in the repository.

```bash
$ git commit
```

**This command:**
- Will open the code editor that is specified in your configuration
- Inside the code editor:
  - A commit message **must** be supplied
  - Lines that start with `#` are comments and will not be recorded
  - Save the file after adding a commit message
  - Close the editor to make the commit
- Then, use `git log` to review the commit you just made!

**Alternative - Quick Commit with Message:**
```bash
$ git commit -m "Your commit message here"
```

**Bypass the editor** by providing the message directly with the `-m` flag.

**💡 Best Practices:**
- Each commit should have a **single focus** (one unit of change)
- Commit messages should clearly describe what changed
- Don't include unrelated changes in the same commit
- Think: "What if all changes in this commit were erased?" - it should only remove one thing

**Example workflow:**
```bash
# 1. Make changes to files
# 2. Stage the changes
$ git add index.html

# 3. Check status
$ git status

# 4. Commit the changes
$ git commit -m "Add header to blog"

# 5. Review the commit
$ git log
```

### ✍️ Commit Messages

#### Writing Good Commit Messages

**✅ DO:**
- Keep messages short (less than 60 characters)
- Explain **what** the commit does (not how or why)
- Use the imperative mood: "Add feature" not "Added feature"
- Think: "This commit will..." and use that to finish your message

**❌ DON'T:**
- Don't explain how the changes are made (that's what `git log -p` is for)
- Don't use the word "and" - if you need it, split into separate commits
- Don't be vague: ❌ "Add changes to app.js" → ✅ "Add event listener to submit button"

**💡 Examples:**

| Bad ❌ | Good ✅ |
|--------|---------|
| "Updated footer" | "Add copyright year to footer" |
| "Fix bug and add feature" | Split into 2 commits |
| "Made changes to CSS" | "Center header content" |

**📝 Adding Details (Optional):**

If you need to explain **why**, add a body after a blank line:

```
Add user authentication

Users requested secure login functionality.
Implements JWT token-based authentication.
Related issue: #123
```

**First line** = shown in `git log --oneline`  
**Body** = shown in full `git log`

**🎯 Pro Tip:** Be consistent with your team's commit message style!

### 🔍 Git Diff

#### Git Diff Recap

The `git diff` command is used to see changes that have been made but haven't been committed yet.

```bash
$ git diff
```

**This command displays:**
- The files that have been modified
- The location of the lines that have been added/removed
- The actual changes that have been made

**What it shows:**
- Lines with `-` (red) = removed/old content
- Lines with `+` (green) = added/new content

**💡 Common Usage:**

```bash
# See changes in working directory (unstaged)
$ git diff

# See changes that are staged for commit
$ git diff --staged

# Compare specific files
$ git diff index.html
```

**Use case:** Review what you've changed before staging or committing files.

### 🚫 Having Git Ignore Files

#### .gitignore Recap

The `.gitignore` file is used to tell Git about files that Git should not track. This file should be placed in the same directory that the `.git` directory is in.

**Why use .gitignore?**
- Prevent accidentally committing unwanted files (e.g., temporary files, build artifacts, credentials)
- Keep your repository clean and focused on source code
- Avoid issues when using `git add .`

**How to use:**

1. Create a `.gitignore` file in your project root
2. List files/patterns you want Git to ignore
3. Git will automatically exclude them from tracking

**Example .gitignore:**
```
# Ignore specific file
project.docx

# Ignore all .log files
*.log

# Ignore all files in build directory
build/

# Ignore all .png files in images folder
images/*.png
```

**🌟 Globbing Patterns:**

| Pattern | Matches | Example |
|---------|---------|---------|
| `*` | 0 or more characters | `*.jpg` → all JPEG files |
| `?` | Exactly 1 character | `file?.txt` → file1.txt, fileA.txt |
| `[abc]` | Matches a, b, or c | `file[123].txt` → file1.txt, file2.txt |
| `**` | Nested directories | `a/**/z` → a/z, a/b/z, a/b/c/z |
| `#` | Comment line | `# This is a comment` |
| Blank lines | Spacing (ignored) | |

**💡 Common Files to Ignore:**
- `.DS_Store` (macOS system file)
- `node_modules/` (dependencies)
- `*.log` (log files)
- `.env` (environment variables/secrets)
- `dist/` or `build/` (compiled output)

# 🏷️ Tagging, Branching and Merging

## 📌 Git Tag

### 🎯 Git Tag 

The `git tag` command is used to add a marker on a specific commit. The tag does not move around as new commits are added.

```bash
git tag -a beta
```

**This command will:**
- Add a tag to the most recent commit
- Add a tag to a specific commit if a SHA is passed

**Example with specific commit:**
```bash
git tag -a v1.0 a87984
```

## 🌿 Git Branching

### 🎯 What is a Branch?

A **branch** is an independent line of development that allows you to work on features or fixes without affecting the main project.

### 📋 Git Branch Commands

#### List All Branches
```bash
git branch
```
- Lists all branch names in the repository
- Active branch is marked with an asterisk `*`

#### Create a New Branch
```bash
git branch sidebar
```
- Creates a new branch called "sidebar"
- **Note:** Creating a branch does NOT switch to it

#### Switch to a Branch
```bash
git checkout sidebar
```
- Switches to the "sidebar" branch
- Removes all files from Working Directory that Git is tracking
- Pulls out all files from the commit that the branch points to

#### Create and Switch in One Command
```bash
git checkout -b new-feature
```
- Creates a new branch and immediately switches to it

#### Delete a Branch
```bash
git branch -d sidebar      # Safe delete (prevents deleting unmerged branches)
git branch -D sidebar      # Force delete (deletes even if unmerged)
```

### 🔍 Viewing Branches in Log

```bash
git log --oneline --decorate
```
- Shows branch information in the commit log
- **HEAD** pointer indicates the current active branch

### ⚠️ Important Notes

- You cannot delete a branch you're currently on
- Git won't let you delete a branch with unique commits (unless you force it with `-D`)
- The **master** branch is the default main branch
- Commits are added to the **current active branch** only
