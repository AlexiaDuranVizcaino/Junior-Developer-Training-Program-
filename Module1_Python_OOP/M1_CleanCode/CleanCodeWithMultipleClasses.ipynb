## 📚 Overview

This notebook covers **class collaboration** and common **code smells** that affect code maintainability. Learn to identify and refactor problematic patterns for better software design.

## 🔴 Code Smell #1: Feature Envy

**Problem:** A method is too interested in another class's data/methods instead of its own.

### ❌ Bad Code - Feature Envy Example

```python
# ❌ BAD: ShoppingCart is too interested in Item's internals
class ShoppingCart:
    def __init__(self):
        self.items = []

    def calculate_total_price(self):
        total = 0
        for item in self.items:
            # Too interested in Item's data!
            total += item.get_price() * item.get_quantity()
        return total

class Item:
    def __init__(self, price, quantity):
        self.price = price
        self.quantity = quantity
    
    def get_price(self):
        return self.price
    
    def get_quantity(self):
        return self.quantity

print("This is Feature Envy - ShoppingCart depends too much on Item's internals")
```

### ✅ Good Code - Refactored

```python
# ✅ GOOD: Each class handles its own responsibility
class ShoppingCart:
    def __init__(self):
        self.items = []

    def calculate_total_price(self):
        # Delegate to Item - Item knows how to calculate its own total
        return sum(item.calculate_total() for item in self.items)

class Item:
    def __init__(self, price, quantity):
        self.price = price
        self.quantity = quantity
    
    def calculate_total(self):
        """Item is responsible for calculating its own total"""
        return self.price * self.quantity

print("✔️ Better: Clear separation of concerns")
```

**📌 Key Point:** Move logic to the class that owns the data. Each class should focus on its own data, not others'.

## 🔴 Code Smell #2: Inappropriate Intimacy

**Problem:** Two classes are too closely interconnected, sharing private details.

### ❌ Bad Code - Inappropriate Intimacy

```python
# ❌ BAD: Library knows too much about Book's internal structure
class Library:
    def __init__(self, book):
        self.book = book

    def print_book_details(self):
        # Library is too intimate with Book's internal structure
        print(f"Title: {self.book.get_title()}")
        print(f"Author: {self.book.get_author()}")

class Book:
    def __init__(self, title, author):
        self.title = title
        self.author = author
    
    def get_title(self):
        return self.title
    
    def get_author(self):
        return self.author

print("This is Inappropriate Intimacy - Too much coupling between classes")
```

### ✅ Good Code - Refactored

```python
# ✅ GOOD: Let Book handle its own representation
class Library:
    def __init__(self, book):
        self.book = book

    def print_book_details(self):
        # Library just asks for details, doesn't know the structure
        print(self.book.get_details())

class Book:
    def __init__(self, title, author):
        self.title = title
        self.author = author
    
    def get_details(self):
        """Book encapsulates how to format its details"""
        return f"Title: {self.title}\nAuthor: {self.author}"

print("✔️ Better: Book encapsulates its own representation")
```

**📌 Key Point:** Encapsulate internal details within the class. Don't let other classes know your internal structure. 🛡️

## 🔴 Code Smell #3: Message Chains

**Problem:** Long chains of method calls to access data across multiple objects.

### ❌ Bad Code - Message Chains

```python
# ❌ BAD: Long message chain
class User:
    def __init__(self, address):
        self.address = address

    def get_address(self):
        return self.address

class Address:
    def __init__(self, zip_code):
        self.zip_code = zip_code

    def get_zip_code(self):
        return self.zip_code

class ZipCode:
    def __init__(self, postal_code):
        self.postal_code = postal_code

    def get_postal_code(self):
        return self.postal_code

# ❌ BAD: Messy chain of calls
user = User(Address(ZipCode("90210")))
postal_code = user.get_address().get_zip_code().get_postal_code()
print(f"Postal Code (Bad Way): {postal_code}")
print("❌ This is a Message Chain - Hard to understand the path!")
```

### ✅ Good Code - Refactored

```python
# ✅ GOOD: Encapsulate the navigation
class User:
    def __init__(self, address):
        self.address = address

    def get_user_postal_code(self):
        """User handles getting its own postal code - no chain needed"""
        return self.address.get_postal_code()

class Address:
    def __init__(self, zip_code):
        self.zip_code = zip_code

    def get_postal_code(self):
        """Address handles getting postal code - hides ZipCode details"""
        return self.zip_code.get_postal_code()

class ZipCode:
    def __init__(self, postal_code):
        self.postal_code = postal_code

    def get_postal_code(self):
        return self.postal_code

# ✅ GOOD: Clean, direct call
user = User(Address(ZipCode("90210")))
postal_code = user.get_user_postal_code()
print(f"Postal Code (Good Way): {postal_code}")
print("✔️ Clean and direct interface")
```

**📌 Key Point:** Create intermediate methods to hide navigation paths. This makes code clearer and easier to maintain. 📬

## 🔴 Code Smell #4: Middle Man

**Problem:** A class exists mainly to delegate to another class without adding value.

### ❌ Bad Code - Middle Man

```python
# ❌ BAD: Controller just delegates without adding value
class Controller:
    def __init__(self, service):
        self.service = service

    def execute(self):
        # Controller doesn't do anything - just passes the call
        self.service.perform_action()

class Service:
    def perform_action(self):
        print("Action performed!")

print("❌ Controller is a Middle Man - It only delegates!")
```

### ✅ Good Code - Refactored

```python
# ✅ GOOD: Remove the unnecessary middle man
class Service:
    def perform_action(self):
        print("Action performed!")

# Direct usage - no unnecessary layers
service = Service()
service.perform_action()

print("✔️ Streamlined and efficient - No unnecessary delegation")
```

**📌 Key Point:** Eliminate unnecessary delegation layers. Keep the design simple and direct. 🔥

## ✨ Key Takeaways

Remember these principles to write clean, maintainable code:

✅ **Single Responsibility** - Each class handles one concern  
✅ **Encapsulation** - Hide internal details, expose only what's needed  
✅ **Clear Interfaces** - Simple, direct method calls  
✅ **Minimal Dependencies** - Reduce coupling between classes

## 🎓 Quick Summary

By identifying and refactoring these code smells, you create better code:

1. **Feature Envy** → Move methods to the class that owns the data
2. **Inappropriate Intimacy** → Encapsulate details within classes
3. **Message Chains** → Create intermediate methods to hide navigation
4. **Middle Man** → Remove unnecessary delegation layers

## 📚 Overview

This unit covers **Abstract Base Classes (ABCs)** in Python, which establish contracts for derived classes. ABCs promote flexibility, consistency, and scalability by defining which methods must be implemented without enforcing specific implementations.

## 🔷 What are Abstract Base Classes?

**Definition:** ABCs define a contract specifying which methods derived classes must implement, providing guidance without enforcing specific implementation.

### ✅ Basic ABC Example

```python
from abc import ABC, abstractmethod

# ✅ Abstract Base Class defining a contract
class PaymentProcessor(ABC):
    @abstractmethod
    def process_payment(self, amount):
        """All payment processors must implement this method"""
        pass

# Concrete implementation
class CreditCardProcessor(PaymentProcessor):
    def process_payment(self, amount):
        print(f"Processing credit card payment of ${amount}")

# Usage
processor = CreditCardProcessor()
processor.process_payment(100)

# ❌ This would fail - can't instantiate abstract class:
# abstract_processor = PaymentProcessor()  # TypeError!

print("✔️ ABCs ensure consistent interface across implementations")
```

**📌 Key Point:** ABCs cannot be instantiated directly. They enforce that derived classes implement specific methods while allowing flexibility in how.

## 🔷 Mixing Abstract and Concrete Methods

**Benefit:** ABCs can provide shared functionality (concrete methods) while enforcing specific behaviors (abstract methods).

### ✅ ABC with Both Method Types

```python
from abc import ABC, abstractmethod

# ABC with both abstract and concrete methods
class Animal(ABC):
    def eat(self):
        """Concrete method - shared by all animals"""
        print("This animal is eating.")
    
    @abstractmethod
    def make_sound(self):
        """Abstract method - must be implemented by each animal"""
        pass

# Concrete implementations
class Dog(Animal):
    def make_sound(self):
        print("Bark!")

class Cat(Animal):
    def make_sound(self):
        print("Meow!")

# Usage
dog = Dog()
dog.eat()          # Uses inherited concrete method
dog.make_sound()   # Uses its own implementation

cat = Cat()
cat.eat()          # Uses inherited concrete method  
cat.make_sound()   # Uses its own implementation

print("✔️ Shared behavior (eat) + enforced customization (make_sound)")
```

**📌 Key Point:** Combine concrete methods for shared functionality with abstract methods for required customization. This reduces code duplication. 🔄

## 🔴 Problem: Rigid Payment System

**Issue:** Tightly coupled class with separate methods for each payment type. Adding new payment methods requires modifying existing code.

### ❌ Bad Code - Rigid Structure

```python
# ❌ BAD: Rigid structure - hard to extend
class PaymentProcessor:
    def pay_with_cash(self, amount):
        print(f"Paying ${amount} with cash")
    
    def pay_with_card(self, amount):
        print(f"Paying ${amount} with credit card")
    
    # Need to add a new payment type? Must modify this class!
    # This violates the Open/Closed Principle

processor = PaymentProcessor()
processor.pay_with_cash(50)
processor.pay_with_card(100)

print("❌ Adding new payment types requires modifying existing code!")
```

### ✅ Good Code - Flexible with ABCs

```python
from abc import ABC, abstractmethod

# ✅ GOOD: Flexible design with ABC
class Payment(ABC):
    @abstractmethod
    def pay(self, amount):
        """Each payment type implements its own payment logic"""
        pass

class CashPayment(Payment):
    def pay(self, amount):
        print(f"Paying ${amount} with cash")

class CreditCardPayment(Payment):
    def pay(self, amount):
        print(f"Paying ${amount} with credit card")

# Easy to add new payment types without modifying existing code!
class MobilePayment(Payment):
    def pay(self, amount):
        print(f"Paying ${amount} with mobile payment")

# Flexible function that works with any Payment type
def process_payment(payment_method, amount):
    payment_method.pay(amount)

# Usage - all payment types work the same way
process_payment(CashPayment(), 100)
process_payment(CreditCardPayment(), 150)
process_payment(MobilePayment(), 200)

print("✔️ Easy to extend with new payment types - Open/Closed Principle!")
```

**📌 Key Point:** ABCs enable the **Open/Closed Principle** - open for extension, closed for modification. Add new types without changing existing code! 🚀

## ✨ Best Practices for ABCs

Follow these principles when designing with Abstract Base Classes:

✅ **Single Responsibility** - Keep ABCs focused on one purpose  
✅ **Design for Change** - Make class hierarchies adaptable for future needs  
✅ **Favor Composition Over Inheritance** - Use ABCs to compose behaviors, avoid deep inheritance chains  
✅ **Consistent Interface** - Ensure all implementations follow the same contract

## 🎓 Quick Summary

**Abstract Base Classes** provide powerful tools for flexible design:

1. **Define Contracts** → Specify required methods without enforcing implementation details
2. **Mix Abstract & Concrete** → Share common functionality while enforcing customization
3. **Enable Extensibility** → Add new types easily without modifying existing code (Open/Closed Principle)
4. **Ensure Consistency** → All implementations follow the same interface

**Benefits:**
- ✅ Flexibility - Easy to extend with new implementations
- ✅ Maintainability - Clear contracts reduce coupling
- ✅ Scalability - Add features without breaking existing code
- ✅ Consistency - Enforce uniform interfaces across implementations
