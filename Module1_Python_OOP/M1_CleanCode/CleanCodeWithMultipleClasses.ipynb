## Unit 1: Class Collaboration and Code Smells

### Overview of Class Collaboration Challenges

Let's dive into the challenges of class collaboration by focusing on four common code smells:

- **Feature Envy:** Occurs when a method in one class is overly interested in methods or data in another class.
- **Inappropriate Intimacy:** Describes a situation where two classes are too closely interconnected, sharing private details.
- **Message Chains:** Refers to sequences of method calls across several objects, indicating a lack of clear abstraction.
- **Middle Man:** Exists when a class mainly delegates its behavior to another class without adding functionality.

Understanding these code smells will enable you to improve your class designs, resulting in cleaner and more maintainable code.

## Feature Envy 👀

**Problem:** A method uses another class's data/methods more than its own.

**Solution:** Move behavior to the class that owns the data.

**Benefits:**
- ✅ Reduces coupling
- ✅ Encapsulation improves
- ✅ Responsibilities align with data

## Inappropriate Intimacy 🔐

**Problem:** One class relies too much on another class's internal details.

**Solution:** Encapsulate and expose a clear public API; push behavior to the owner class.

**Key Difference from Feature Envy:**
- Inappropriate Intimacy = significant intertwining between two classes
- Feature Envy = method's excessive interest in another's data/methods

**Benefits:**
- ✅ Better encapsulation
- ✅ Clearer contracts between classes
- ✅ Easier to modify internals

```python
# Example (before refactor): Feature Envy
class ShoppingCart:
    def __init__(self):
        self.items = []

    def calculate_total_price(self):
        total = 0
        for item in self.items:
            # Accessing Item internals too much
            total += item.get_price() * item.get_quantity()
        return total

class Item:
    def __init__(self, price, quantity):
        self.price = price
        self.quantity = quantity

    def get_price(self):
        return self.price

    def get_quantity(self):
        return self.quantity

# Usage
cart = ShoppingCart()
cart.items.append(Item(10, 2))
print('Before refactor total:', cart.calculate_total_price())
```

```python
# Example (after refactor): Move responsibility to Item
class ShoppingCart:
    def __init__(self):
        self.items = []

    def calculate_total_price(self):
        total = 0
        for item in self.items:
            total += item.calculate_total()
        return total

class Item:
    def __init__(self, price, quantity):
        self.price = price
        self.quantity = quantity

    def calculate_total(self):
        return self.price * self.quantity

# Usage
cart = ShoppingCart()
cart.items.append(Item(10, 2))
print('After refactor total:', cart.calculate_total_price())
```

## Message Chains 🔗

**Problem:** Code traverses multiple objects to reach desired method: `obj.get_a().get_b().get_c()`.

**Solution:** Encapsulate access in a single method or use a façade.

**Benefits:**
- ✅ Simpler, more readable code
- ✅ Reduces coupling to intermediate objects
- ✅ Changes in chain don't affect caller

```python
# Before: Library accesses Book internals
class Library:
    def __init__(self, book):
        self.book = book

    def print_book_details(self):
        # Accessing book getters and assembling details here
        print(f"Title: {self.book.get_title()}")
        print(f"Author: {self.book.get_author()}")

class Book:
    def __init__(self, title, author):
        self.title = title
        self.author = author

    def get_title(self):
        return self.title

    def get_author(self):
        return self.author

# Usage
book = Book('1984', 'George Orwell')
lib = Library(book)
lib.print_book_details()
```

```python
# After: Book exposes a clear API for its representation
class Library:
    def __init__(self, book):
        self.book = book

    def print_book_details(self):
        print(self.book.get_details())

class Book:
    def __init__(self, title, author):
        self.title = title
        self.author = author

    def get_details(self):
        return f"Title: {self.title}
Author: {self.author}"

# Usage
book = Book('1984', 'George Orwell')
lib = Library(book)
lib.print_book_details()
```

## Message Chains

Message Chains happen when code traverses several objects to get data (obj.get_a().get_b().get_c()). Long chains indicate missing abstraction — provide a direct method or a façade.

```python
# Before: long message chain
class ZipCode:
    def __init__(self, postal_code):
        self.postal_code = postal_code

    def get_postal_code(self):
        return self.postal_code

class Address:
    def __init__(self, zip_code):
        self.zip_code = zip_code

    def get_zip_code(self):
        return self.zip_code

class User:
    def __init__(self, address):
        self.address = address

    def get_address(self):
        return self.address

# Usage
user = User(Address(ZipCode('90210')))
postal = user.get_address().get_zip_code().get_postal_code()
print('Before chain postal code:', postal)
```

```python
# After: encapsulate access with a single method
class ZipCode:
    def __init__(self, postal_code):
        self.postal_code = postal_code

    def get_postal_code(self):
        return self.postal_code

class Address:
    def __init__(self, zip_code):
        self.zip_code = zip_code

    def get_postal_code(self):
        return self.zip_code.get_postal_code()

class User:
    def __init__(self, address):
        self.address = address

    def get_user_postal_code(self):
        return self.address.get_postal_code()

# Usage
user = User(Address(ZipCode('90210')))
print('After refactor postal code:', user.get_user_postal_code())
```

## Middle Man 🤝

**Problem:** A class exists only to delegate; it doesn't add functionality.

**Solution:** Remove the middle man or assign real responsibilities.

**Benefits:**
- ✅ Simpler, more direct design
- ✅ Fewer unnecessary abstraction layers
- ✅ Easier to understand flow

```python
# Before: Controller is mostly a middle man
class Controller:
    def __init__(self, service):
        self.service = service

    def execute(self):
        self.service.perform_action()

class Service:
    def perform_action(self):
        print('Action performed')

# Usage
svc = Service()
ctrl = Controller(svc)
ctrl.execute()
```

```python
# After: remove unnecessary delegation or enrich Controller if needed
class Service:
    def perform_action(self):
        print('Action performed')

# Usage - call service directly
svc = Service()
svc.perform_action()
```
