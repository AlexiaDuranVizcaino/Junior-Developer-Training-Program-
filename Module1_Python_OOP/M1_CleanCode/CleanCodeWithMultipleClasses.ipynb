## Unit 1: Code Smells & Class Collaboration

## 📚 Overview

This notebook covers **class collaboration** and common **code smells** that affect code maintainability. Learn to identify and refactor problematic patterns for better software design.

## 🔴 Code Smell #1: Feature Envy

**Problem:** A method is too interested in another class's data/methods instead of its own.

### ❌ Bad Code - Feature Envy Example

```python
# ❌ BAD: ShoppingCart is too interested in Item's internals
class ShoppingCart:
    def __init__(self):
        self.items = []

    def calculate_total_price(self):
        total = 0
        for item in self.items:
            # Too interested in Item's data!
            total += item.get_price() * item.get_quantity()
        return total

class Item:
    def __init__(self, price, quantity):
        self.price = price
        self.quantity = quantity
    
    def get_price(self):
        return self.price
    
    def get_quantity(self):
        return self.quantity

print("This is Feature Envy - ShoppingCart depends too much on Item's internals")
```

### ✅ Good Code - Refactored

```python
# ✅ GOOD: Each class handles its own responsibility
class ShoppingCart:
    def __init__(self):
        self.items = []

    def calculate_total_price(self):
        # Delegate to Item - Item knows how to calculate its own total
        return sum(item.calculate_total() for item in self.items)

class Item:
    def __init__(self, price, quantity):
        self.price = price
        self.quantity = quantity
    
    def calculate_total(self):
        """Item is responsible for calculating its own total"""
        return self.price * self.quantity

print("✔️ Better: Clear separation of concerns")
```

**📌 Key Point:** Move logic to the class that owns the data. Each class should focus on its own data, not others'.

## 🔴 Code Smell #2: Inappropriate Intimacy

**Problem:** Two classes are too closely interconnected, sharing private details.

### ❌ Bad Code - Inappropriate Intimacy

```python
# ❌ BAD: Library knows too much about Book's internal structure
class Library:
    def __init__(self, book):
        self.book = book

    def print_book_details(self):
        # Library is too intimate with Book's internal structure
        print(f"Title: {self.book.get_title()}")
        print(f"Author: {self.book.get_author()}")

class Book:
    def __init__(self, title, author):
        self.title = title
        self.author = author
    
    def get_title(self):
        return self.title
    
    def get_author(self):
        return self.author

print("This is Inappropriate Intimacy - Too much coupling between classes")
```

### ✅ Good Code - Refactored

```python
# ✅ GOOD: Let Book handle its own representation
class Library:
    def __init__(self, book):
        self.book = book

    def print_book_details(self):
        # Library just asks for details, doesn't know the structure
        print(self.book.get_details())

class Book:
    def __init__(self, title, author):
        self.title = title
        self.author = author
    
    def get_details(self):
        """Book encapsulates how to format its details"""
        return f"Title: {self.title}\nAuthor: {self.author}"

print("✔️ Better: Book encapsulates its own representation")
```

**📌 Key Point:** Encapsulate internal details within the class. Don't let other classes know your internal structure. 🛡️

## 🔴 Code Smell #3: Message Chains

**Problem:** Long chains of method calls to access data across multiple objects.

### ❌ Bad Code - Message Chains

```python
# ❌ BAD: Long message chain
class User:
    def __init__(self, address):
        self.address = address

    def get_address(self):
        return self.address

class Address:
    def __init__(self, zip_code):
        self.zip_code = zip_code

    def get_zip_code(self):
        return self.zip_code

class ZipCode:
    def __init__(self, postal_code):
        self.postal_code = postal_code

    def get_postal_code(self):
        return self.postal_code

# ❌ BAD: Messy chain of calls
user = User(Address(ZipCode("90210")))
postal_code = user.get_address().get_zip_code().get_postal_code()
print(f"Postal Code (Bad Way): {postal_code}")
print("❌ This is a Message Chain - Hard to understand the path!")
```

### ✅ Good Code - Refactored

```python
# ✅ GOOD: Encapsulate the navigation
class User:
    def __init__(self, address):
        self.address = address

    def get_user_postal_code(self):
        """User handles getting its own postal code - no chain needed"""
        return self.address.get_postal_code()

class Address:
    def __init__(self, zip_code):
        self.zip_code = zip_code

    def get_postal_code(self):
        """Address handles getting postal code - hides ZipCode details"""
        return self.zip_code.get_postal_code()

class ZipCode:
    def __init__(self, postal_code):
        self.postal_code = postal_code

    def get_postal_code(self):
        return self.postal_code

# ✅ GOOD: Clean, direct call
user = User(Address(ZipCode("90210")))
postal_code = user.get_user_postal_code()
print(f"Postal Code (Good Way): {postal_code}")
print("✔️ Clean and direct interface")
```

**📌 Key Point:** Create intermediate methods to hide navigation paths. This makes code clearer and easier to maintain. 📬

## 🔴 Code Smell #4: Middle Man

**Problem:** A class exists mainly to delegate to another class without adding value.

### ❌ Bad Code - Middle Man

```python
# ❌ BAD: Controller just delegates without adding value
class Controller:
    def __init__(self, service):
        self.service = service

    def execute(self):
        # Controller doesn't do anything - just passes the call
        self.service.perform_action()

class Service:
    def perform_action(self):
        print("Action performed!")

print("❌ Controller is a Middle Man - It only delegates!")
```

### ✅ Good Code - Refactored

```python
# ✅ GOOD: Remove the unnecessary middle man
class Service:
    def perform_action(self):
        print("Action performed!")

# Direct usage - no unnecessary layers
service = Service()
service.perform_action()

print("✔️ Streamlined and efficient - No unnecessary delegation")
```

**📌 Key Point:** Eliminate unnecessary delegation layers. Keep the design simple and direct. 🔥

## ✨ Key Takeaways

Remember these principles to write clean, maintainable code:

✅ **Single Responsibility** - Each class handles one concern  
✅ **Encapsulation** - Hide internal details, expose only what's needed  
✅ **Clear Interfaces** - Simple, direct method calls  
✅ **Minimal Dependencies** - Reduce coupling between classes

## 🎓 Quick Summary

By identifying and refactoring these code smells, you create better code:

1. **Feature Envy** → Move methods to the class that owns the data
2. **Inappropriate Intimacy** → Encapsulate details within classes
3. **Message Chains** → Create intermediate methods to hide navigation
4. **Middle Man** → Remove unnecessary delegation layers

## 📚 Overview

This unit covers **Abstract Base Classes (ABCs)** in Python, which establish contracts for derived classes. ABCs promote flexibility, consistency, and scalability by defining which methods must be implemented without enforcing specific implementations.

## 🔷 What are Abstract Base Classes?

**Definition:** ABCs define a contract specifying which methods derived classes must implement, providing guidance without enforcing specific implementation.

### ✅ Basic ABC Example

```python
from abc import ABC, abstractmethod

# ✅ Abstract Base Class defining a contract
class PaymentProcessor(ABC):
    @abstractmethod
    def process_payment(self, amount):
        """All payment processors must implement this method"""
        pass

# Concrete implementation
class CreditCardProcessor(PaymentProcessor):
    def process_payment(self, amount):
        print(f"Processing credit card payment of ${amount}")

# Usage
processor = CreditCardProcessor()
processor.process_payment(100)

# ❌ This would fail - can't instantiate abstract class:
# abstract_processor = PaymentProcessor()  # TypeError!

print("✔️ ABCs ensure consistent interface across implementations")
```

**📌 Key Point:** ABCs cannot be instantiated directly. They enforce that derived classes implement specific methods while allowing flexibility in how.

## 🔷 Mixing Abstract and Concrete Methods

**Benefit:** ABCs can provide shared functionality (concrete methods) while enforcing specific behaviors (abstract methods).

### ✅ ABC with Both Method Types

```python
from abc import ABC, abstractmethod

# ABC with both abstract and concrete methods
class Animal(ABC):
    def eat(self):
        """Concrete method - shared by all animals"""
        print("This animal is eating.")
    
    @abstractmethod
    def make_sound(self):
        """Abstract method - must be implemented by each animal"""
        pass

# Concrete implementations
class Dog(Animal):
    def make_sound(self):
        print("Bark!")

class Cat(Animal):
    def make_sound(self):
        print("Meow!")

# Usage
dog = Dog()
dog.eat()          # Uses inherited concrete method
dog.make_sound()   # Uses its own implementation

cat = Cat()
cat.eat()          # Uses inherited concrete method  
cat.make_sound()   # Uses its own implementation

print("✔️ Shared behavior (eat) + enforced customization (make_sound)")
```

**📌 Key Point:** Combine concrete methods for shared functionality with abstract methods for required customization. This reduces code duplication. 🔄

## 🔴 Problem: Rigid Payment System

**Issue:** Tightly coupled class with separate methods for each payment type. Adding new payment methods requires modifying existing code.

### ❌ Bad Code - Rigid Structure

```python
# ❌ BAD: Rigid structure - hard to extend
class PaymentProcessor:
    def pay_with_cash(self, amount):
        print(f"Paying ${amount} with cash")
    
    def pay_with_card(self, amount):
        print(f"Paying ${amount} with credit card")
    
    # Need to add a new payment type? Must modify this class!
    # This violates the Open/Closed Principle

processor = PaymentProcessor()
processor.pay_with_cash(50)
processor.pay_with_card(100)

print("❌ Adding new payment types requires modifying existing code!")
```

### ✅ Good Code - Flexible with ABCs

```python
from abc import ABC, abstractmethod

# ✅ GOOD: Flexible design with ABC
class Payment(ABC):
    @abstractmethod
    def pay(self, amount):
        """Each payment type implements its own payment logic"""
        pass

class CashPayment(Payment):
    def pay(self, amount):
        print(f"Paying ${amount} with cash")

class CreditCardPayment(Payment):
    def pay(self, amount):
        print(f"Paying ${amount} with credit card")

# Easy to add new payment types without modifying existing code!
class MobilePayment(Payment):
    def pay(self, amount):
        print(f"Paying ${amount} with mobile payment")

# Flexible function that works with any Payment type
def process_payment(payment_method, amount):
    payment_method.pay(amount)

# Usage - all payment types work the same way
process_payment(CashPayment(), 100)
process_payment(CreditCardPayment(), 150)
process_payment(MobilePayment(), 200)

print("✔️ Easy to extend with new payment types - Open/Closed Principle!")
```

**📌 Key Point:** ABCs enable the **Open/Closed Principle** - open for extension, closed for modification. Add new types without changing existing code! 🚀

## ✨ Best Practices for ABCs

Follow these principles when designing with Abstract Base Classes:

✅ **Single Responsibility** - Keep ABCs focused on one purpose  
✅ **Design for Change** - Make class hierarchies adaptable for future needs  
✅ **Favor Composition Over Inheritance** - Use ABCs to compose behaviors, avoid deep inheritance chains  
✅ **Consistent Interface** - Ensure all implementations follow the same contract

## 🎓 Quick Summary

**Abstract Base Classes** provide powerful tools for flexible design:

1. **Define Contracts** → Specify required methods without enforcing implementation details
2. **Mix Abstract & Concrete** → Share common functionality while enforcing customization
3. **Enable Extensibility** → Add new types easily without modifying existing code (Open/Closed Principle)
4. **Ensure Consistency** → All implementations follow the same interface

**Benefits:**
- ✅ Flexibility - Easy to extend with new implementations
- ✅ Maintainability - Clear contracts reduce coupling
- ✅ Scalability - Add features without breaking existing code
- ✅ Consistency - Enforce uniform interfaces across implementations

## Unit 3: Dependency Management Between Classes

---

## 📚 Overview

This unit focuses on **managing dependencies** between classes to create maintainable, testable, and flexible code. Learn how to decouple classes using dependency injection and the Dependency Inversion Principle.

## 🔷 Understanding Dependencies

**Definition:** Dependencies occur when one class relies on another class's functionality. Tight coupling makes code rigid and hard to test.

### ❌ Tightly Coupled Code

```python
# ❌ BAD: Car directly creates its Engine - tight coupling
class Engine:
    def start(self):
        print("Engine starting...")

class Car:
    def __init__(self):
        self.engine = Engine()  # ❌ Direct dependency - hard to change or test
    
    def start(self):
        self.engine.start()

# Problems:
# - Can't easily swap engine types
# - Hard to test Car without real Engine
# - Changes to Engine affect Car

car = Car()
car.start()

print("❌ Tight coupling - Car is stuck with one Engine type!")
```

**📌 Key Point:** Direct instantiation creates tight coupling. Changes in one class ripple through dependent classes, making code fragile and hard to test.

## 🔷 Dependency Injection Solution

**Benefit:** Pass dependencies as parameters instead of creating them internally. This decouples classes and improves testability.

### ✅ Loosely Coupled with Dependency Injection

```python
# ✅ GOOD: Inject the dependency from outside
class Engine:
    def start(self):
        print("Engine starting...")

class ElectricEngine:
    def start(self):
        print("Electric motor powering up...")

class Car:
    def __init__(self, engine):  # ✅ Dependency injection via constructor
        self.engine = engine
    
    def start(self):
        self.engine.start()

# Now we can easily swap engines!
gas_car = Car(Engine())
electric_car = Car(ElectricEngine())

gas_car.start()
electric_car.start()

print("✔️ Flexible - Easy to swap implementations and test!")
```

**📌 Key Point:** Dependency Injection allows flexibility and makes testing easier by enabling mock objects. The class doesn't control what it depends on. 💉

## 🔷 Dependency Inversion Principle (DIP)

**Core Idea:** High-level modules shouldn't depend on low-level modules. Both should depend on abstractions.

### ✅ DIP with Abstract Base Classes

```python
from abc import ABC, abstractmethod

# ✅ Abstract interface - the "contract"
class Engine(ABC):
    @abstractmethod
    def start(self):
        """All engines must implement start()"""
        pass

# Concrete implementations
class GasEngine(Engine):
    def start(self):
        print("Gas engine starting...")

class ElectricEngine(Engine):
    def start(self):
        print("Electric motor powering up...")

# Car depends on abstraction, not concrete classes
class Car:
    def __init__(self, engine: Engine):  # ✅ Depends on abstraction
        self.engine = engine
    
    def start(self):
        self.engine.start()

# Works with any Engine implementation
gas_car = Car(GasEngine())
electric_car = Car(ElectricEngine())

gas_car.start()
electric_car.start()

print("✔️ DIP: Car depends on Engine abstraction, not specific types!")
```

**📌 Key Point:** Depend on abstractions (interfaces/ABCs), not concrete implementations. This makes code flexible and extensible. 🔄

## 🔷 Factory Pattern for Dependency Management

**Benefit:** Centralize object creation logic, further reducing coupling and making code easier to maintain.

### ✅ Using Factory Pattern

```python
from abc import ABC, abstractmethod

class Engine(ABC):
    @abstractmethod
    def start(self):
        pass

class GasEngine(Engine):
    def start(self):
        print("Gas engine starting...")

class ElectricEngine(Engine):
    def start(self):
        print("Electric motor powering up...")

# ✅ Factory encapsulates creation logic
class EngineFactory:
    @staticmethod
    def create_engine(engine_type: str) -> Engine:
        if engine_type == "gas":
            return GasEngine()
        elif engine_type == "electric":
            return ElectricEngine()
        else:
            raise ValueError(f"Unknown engine type: {engine_type}")

class Car:
    def __init__(self, engine_type: str):
        # ✅ Delegate creation to factory
        self.engine = EngineFactory.create_engine(engine_type)
    
    def start(self):
        self.engine.start()

# Clean usage - just specify type
gas_car = Car("gas")
electric_car = Car("electric")

gas_car.start()
electric_car.start()

print("✔️ Factory pattern centralizes creation - easy to add new engine types!")
```

**📌 Key Point:** Factory Pattern separates object creation from usage. Adding new types doesn't require changing existing code. 🏭

## ✨ Best Practices for Dependency Management

Follow these principles for clean, maintainable code:

✅ **Use Dependency Injection** - Pass dependencies via constructor/parameters  
✅ **Depend on Abstractions** - Use ABCs/interfaces instead of concrete classes  
✅ **Apply Factory Pattern** - Centralize object creation logic  
✅ **Follow DIP** - High-level modules depend on abstractions, not details  
✅ **Enable Testing** - Loose coupling makes unit testing easier with mocks

## 🎓 Quick Summary

**Dependency Management** keeps code flexible and testable:

1. **Avoid Tight Coupling** → Don't create dependencies inside classes
2. **Use Dependency Injection** → Pass dependencies as parameters
3. **Apply DIP** → Depend on abstractions (ABCs), not concrete classes
4. **Use Factory Pattern** → Centralize and encapsulate object creation
5. **Enable Testability** → Loose coupling allows easy mocking for tests

**Benefits:**
- ✅ Flexibility - Easy to swap implementations
- ✅ Testability - Can inject mocks/stubs for unit tests
- ✅ Maintainability - Changes don't ripple through the codebase
- ✅ Scalability - Add new implementations without breaking existing code

**Remember:** Dependencies are managed through **injection** and **abstraction**, not through direct instantiation! 🚀

# 🎯 Clean Code with Multiple Classes
## Unit 4: Polymorphism in Practice

---

## 📚 What is Polymorphism?

**Definition:** Polymorphism allows writing code that works with objects of different classes in a unified manner. Python's **duck typing** treats objects by their behavior, not their type.

### ✅ Benefits: Unified Payment Processing

```python
from abc import ABC, abstractmethod

# Define abstract base class
class Payment(ABC):
    @abstractmethod
    def pay(self):
        pass

class CreditCardPayment(Payment):
    def pay(self):
        print("Processing credit card payment.")

class PayPalPayment(Payment):
    def pay(self):
        print("Processing PayPal payment.")

# ✅ Polymorphic function - works with ANY Payment type
def process_payment(payment):
    payment.pay()  # Duck typing - no type checking needed!

# Usage
payment = CreditCardPayment()
process_payment(payment)

print("✔️ One interface, multiple implementations!")
```

**📌 Key Point:** Polymorphism reduces code duplication. Add new payment types by implementing the `pay()` method - no changes to existing code!

## 🔴 Problem: Rigid Code Without Polymorphism

**Issue:** Lengthy conditional logic to handle different object types - difficult to modify and extend.

### ❌ Bad Approach

```python
# ❌ WITHOUT polymorphism - rigid conditional logic
def process_payment_details(payment_method):
    if isinstance(payment_method, CreditCardPayment):
        # Process credit card payment
        print("Processing credit card...")
    elif isinstance(payment_method, PayPalPayment):
        # Process PayPal payment
        print("Processing PayPal...")
    # More conditions for each new type...

print("❌ Must modify this function for every new payment type!")
```

### ✅ Solution: Using Polymorphism

```python
# ✅ WITH polymorphism - clean and extensible
def process_payment(payment):
    payment.pay()  # Works for any Payment type!

print("✔️ No conditionals needed - just call the method!")
```

## ✨ Best Practices for Polymorphic Design

Follow these principles when implementing polymorphism:

✅ **Embrace Duck Typing** - Trust objects implement necessary methods without explicit type checks  
✅ **Favor Composition Over Inheritance** - Use composition to share behavior without rigid inheritance chains  
✅ **Keep Methods Consistent** - Ensure polymorphic methods have consistent signatures and behaviors  
✅ **Use Type Hints Judiciously** - Don't over-constrain with type hints; maintain Python's flexibility  
✅ **Design Clear Abstractions** - Abstract common behavior properly to avoid redundant methods

## 🎓 Quick Summary

**Polymorphism** enables flexible, maintainable code:

1. **Unified Interface** → Same method works for different types
2. **Duck Typing** → Objects treated by behavior, not type
3. **Open/Closed Principle** → Extend without modifying existing code
4. **Eliminate Conditionals** → Replace type-checking with polymorphic methods

**Benefits:** Flexibility, maintainability, scalability, cleaner code! 🚀

## Unit 5: Exception Handling Across Classes

---

## 📚 Overview

This unit covers **exception handling** across multiple classes in Python. Proper exception management prevents error propagation, enhances reliability, and maintains code maintainability without cluttering business logic.

## 🔴 Common Problems in Exception Handling

**Issues when exceptions span multiple classes:**

1. **Loss of Exception Context** - Catching and re-raising without adequate information makes diagnosis challenging
2. **Diminished Readability** - Complex exception handling mixed with business logic obscures the main purpose
3. **Tight Coupling** - Poor exception design can violate loose coupling and high cohesion principles

### ❌ Poor Exception Handling

```python
# ❌ BAD: Exception handling without context
class DataService:
    def fetch_data(self):
        try:
            # Some operation that might fail
            raise ValueError("Data fetch failed")
        except ValueError:
            # Lost context - what went wrong?
            raise Exception("Error occurred")

# Problem: Original error details are lost!
print("❌ Poor exception handling - context is lost!")
```

## ✅ Best Practices for Multi-Class Exception Handling

Follow these principles for robust exception management:

1. **Propagate Exceptions with Context** - Use `raise ... from ...` to preserve original exception context
2. **Use Python's Exception Hierarchy** - Leverage built-in exceptions and create custom ones for specific scenarios
3. **Separate Exception Handling from Core Logic** - Keep business logic clean and readable

### ✅ Good Exception Handling

```python
# ✅ GOOD: Preserving exception context with 'raise ... from ...'
class DataAccessException(Exception):
    pass

class DataService:
    def fetch_data(self):
        try:
            # Simulating external service error
            raise ConnectionError("API connection failed")
        except ConnectionError as e:
            # Preserve original exception context
            raise DataAccessException("Failed to retrieve data from external service") from e

# Benefits:
# - Original error is preserved in __cause__ attribute
# - Better traceback for debugging
# - High-level abstraction for application layer

print("✔️ Exception context preserved for debugging!")
```

**📌 Key Point:** Using `raise ... from ...` preserves the original exception in the `__cause__` attribute, providing complete traceback for debugging while maintaining clean abstraction layers.

## 🔷 Custom Exception Classes

**Benefit:** Custom exceptions encapsulate domain-specific error conditions with meaningful context and syntax.

### ✅ Custom Exception Hierarchy

```python
# ✅ Creating a custom exception hierarchy
class APIServiceException(Exception):
    """Base exception for API-related errors."""
    pass

class DataRetrievalException(APIServiceException):
    """Raised when data retrieval fails."""
    def __init__(self, message, cause=None):
        super().__init__(message)
        self.cause = cause

class AuthenticationException(APIServiceException):
    """Raised when authentication fails."""
    pass

# Usage Example
def fetch_user_data(user_id):
    try:
        # Simulating authentication failure
        raise ValueError("Invalid credentials")
    except ValueError as e:
        raise AuthenticationException("User authentication failed") from e

# Benefits:
# - Clear exception hierarchy
# - Domain-specific error types
# - Easier to catch and handle specific errors
# - Better code organization

print("✔️ Custom exceptions provide domain-specific error handling!")
```

**📌 Key Point:** Specific exceptions (like `DataRetrievalException`) inherit from broader categories (`APIServiceException`) to handle domain-specific errors. This promotes informative debugging and clean abstraction.

## 🔷 Exception Propagation Across Service Layers

**Goal:** Maintain distinct boundaries between application layers while preserving context-specific error information.

### ✅ Multi-Layer Exception Handling Example

```python
# ✅ Exception propagation through service layers
class InventoryException(Exception):
    """Exception for inventory-related errors."""
    pass

class OrderProcessingException(Exception):
    """Exception for order processing errors."""
    pass

class InventoryService:
    def reserve_items(self, items):
        if not items:
            raise InventoryException("No items in the order to reserve.")
        # Item reservation logic...
        print(f"Reserved {len(items)} items")

class OrderService:
    def __init__(self, inventory_service):
        self.inventory_service = inventory_service
    
    def process_order(self, order):
        try:
            self.inventory_service.reserve_items(order.get_items())
            print("Order processed successfully!")
        except InventoryException as e:
            # Re-raise with order-specific context
            raise OrderProcessingException("Failed to reserve items") from e

# Usage
class Order:
    def __init__(self, items):
        self.items = items
    
    def get_items(self):
        return self.items

inventory = InventoryService()
order_service = OrderService(inventory)

# Successful order
order1 = Order(["item1", "item2"])
order_service.process_order(order1)

# This would fail:
# order2 = Order([])
# order_service.process_order(order2)  # Raises OrderProcessingException

print("✔️ Clean separation between layers with context preservation!")
```

**📌 Key Point:** The `OrderService` catches `InventoryException` and re-raises it as `OrderProcessingException` with additional context. This maintains distinct boundaries between application layers without losing the original error information.

## 🔷 Keeping Exception Handling Separate from Business Logic

**Principle:** Separate error handling from core logic to enhance readability and maintainability.

### ❌ Mixed Logic and Exception Handling

```python
# ❌ BAD: Exception handling mixed with business logic
def process_payment_bad(amount):
    try:
        if amount <= 0:
            raise ValueError("Invalid amount")
        # Business logic mixed with error handling
        print(f"Processing ${amount}")
        # More validation...
        if amount > 10000:
            raise ValueError("Amount too large")
        # More processing...
        return True
    except ValueError as e:
        print(f"Error: {e}")
        return False

print("❌ Hard to read - business logic obscured by error handling")
```

### ✅ Separated Concerns

```python
# ✅ GOOD: Separate validation and business logic
class PaymentValidator:
    @staticmethod
    def validate_amount(amount):
        if amount <= 0:
            raise ValueError("Invalid amount")
        if amount > 10000:
            raise ValueError("Amount too large")

class PaymentProcessor:
    def __init__(self, validator):
        self.validator = validator
    
    def process_payment(self, amount):
        # Validation happens separately
        self.validator.validate_amount(amount)
        
        # Clean business logic
        print(f"Processing ${amount}")
        return True

# Usage with exception handling at higher level
validator = PaymentValidator()
processor = PaymentProcessor(validator)

try:
    processor.process_payment(500)
    print("✔️ Payment processed!")
except ValueError as e:
    print(f"Payment failed: {e}")

print("✔️ Clear separation: validation, business logic, and error handling!")
```

**📌 Key Point:** Separating validation, business logic, and error handling makes code more readable, testable, and maintainable. Each component has a single, clear responsibility.
