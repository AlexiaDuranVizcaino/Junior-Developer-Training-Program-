# 🎯 Applying Clean Code Principles in Python
## Unit 1: The DRY Principle (Don't Repeat Yourself)

---

## 📚 Overview

This unit covers the **DRY principle**, a fundamental concept in clean coding. Learn to identify and eliminate code repetition through strategic refactoring techniques that improve maintainability and reduce errors.

## 🔷 Understanding the Problem: Why Code Repetition is Bad

**Repetitive code introduces several critical issues:**

1. **Code Bloat** - Unnecessarily increases codebase size, making navigation harder
2. **Risk of Inconsistencies** - Similar logic scattered across files becomes out of sync during updates
3. **Maintenance Challenges** - Updates require changes in multiple places, increasing error likelihood

### 💡 Key Impact
Redundant code makes it difficult to ensure consistent changes across the entire codebase, leading to bugs and technical debt.

## 🛠️ DRY Strategies

To adhere to the DRY principle, use these key refactoring techniques:

✅ **Extracting Function** - Move repeated logic into a reusable function  
✅ **Extracting Variable** - Consolidate repeated expressions into named variables  
✅ **Replace Temp with Function** - Compute values on demand instead of storing in temporary variables

## 🔴 Strategy #1: Extracting Function

**Problem:** Duplicated logic for calculating item totals across different shipping methods.

### ❌ Bad Code - Repeated Logic

```python
# ❌ BAD: Duplicated item total calculation logic
def calculate_click_and_collect_total(order):
    items_total = 0
    for item in order.get_items():
        items_total += item.get_price() * item.get_quantity()
    shipping_cost = 0 if items_total > 100 else 5
    return items_total + shipping_cost + order.get_tax()

def calculate_post_shipment_total(order, is_express):
    items_total = 0
    for item in order.get_items():
        items_total += item.get_price() * item.get_quantity()  # Duplicated!
    shipping_cost = items_total * 0.1 if is_express else items_total * 0.05
    return items_total + shipping_cost + order.get_tax()

print("❌ Repeated logic - hard to maintain and error-prone")
```

### ✅ Good Code - Extracted Function

```python
# ✅ GOOD: Shared logic extracted into reusable function
def calculate_items_total(order):
    """Centralized logic for calculating item totals"""
    items_total = 0
    for item in order.get_items():
        items_total += item.get_price() * item.get_quantity()
    return items_total

def calculate_click_and_collect_total(order):
    items_total = calculate_items_total(order)  # Reuse!
    shipping_cost = 0 if items_total > 100 else 5
    return items_total + shipping_cost + order.get_tax()

def calculate_post_shipment_total(order, is_express):
    items_total = calculate_items_total(order)  # Reuse!
    shipping_cost = items_total * 0.1 if is_express else items_total * 0.05
    return items_total + shipping_cost + order.get_tax()

print("✔️ Clean and maintainable - logic centralized in one place")
```

**📌 Key Point:** Extracting repeated logic into a dedicated function promotes code reuse, simplifies maintenance, and ensures consistency across updates. Changes only need to be made in one place! 🎯

## 🔴 Strategy #2: Extracting Variable

**Problem:** Repeated expressions and unclear calculations make code hard to understand and modify.

### ❌ Bad Code - Scattered Discount Calculations

```python
# ❌ BAD: Unclear calculations without named variables
def apply_discount(price, customer):
    loyalty_discount_rate = customer.get_loyalty_level() * 0.02
    price *= (1 - loyalty_discount_rate)
    # Additional discounts
    seasonal_discount_rate = 0.10
    price *= (1 - seasonal_discount_rate)  # What does this represent?
    return price

# Issues:
# - Hard to understand what each calculation does
# - Discount application steps are unclear
# - Difficult to modify or debug

print("❌ Unclear calculations - what's happening to the price?")
```

### ✅ Good Code - Extracted Variables with Clear Names

```python
# ✅ GOOD: Named variables make each step crystal clear
def apply_discount(price, customer):
    # Extract discount rates into named variables
    loyalty_discount_rate = customer.get_loyalty_level() * 0.02
    seasonal_discount_rate = 0.10
    
    # Each calculation step is clearly named
    loyalty_adjusted_price = price * (1 - loyalty_discount_rate)
    final_price = loyalty_adjusted_price * (1 - seasonal_discount_rate)
    
    return final_price

# Benefits:
# - Clear sequence of operations
# - Easy to understand each discount phase
# - Simple to modify or add new discount types
# - Self-documenting code

print("✔️ Clear and readable - each step has a meaningful name")
```

**📌 Key Point:** Extracting variables with descriptive names improves code clarity by making the purpose and sequence of calculations explicit. This makes the code self-documenting and easier to maintain. 📝

## 🔴 Strategy #3: Replace Temp with Function

**Problem:** Temporary variables used in multiple places cause duplicated logic and make code harder to maintain.

### ❌ Bad Code - Duplicated Temporary Variable

```python
# ❌ BAD: Duplicated temp variable logic
from datetime import datetime, timedelta

def is_eligible_for_discount(customer):
    # Temporary variable duplicated in another function
    new_customer = customer.get_sign_up_date() > datetime.now() - timedelta(days=90)
    return new_customer and len(customer.get_purchase_history()) > 5

def is_eligible_for_loyalty_program(customer):
    # Same logic repeated here!
    new_customer = customer.get_sign_up_date() > datetime.now() - timedelta(days=90)
    return new_customer or customer.get_loyalty_level() > 3

# Problems:
# - Logic duplicated across functions
# - Changes require updates in multiple places
# - Inconsistency risk if one is updated but not the other

print("❌ Duplicated temporary variable - maintenance nightmare")
```

### ✅ Good Code - Replaced with Reusable Function

```python
# ✅ GOOD: Extract logic into a dedicated function
from datetime import datetime, timedelta

def is_new_customer(customer):
    """Centralized logic for determining if customer is new"""
    return customer.get_sign_up_date() > datetime.now() - timedelta(days=90)

def is_eligible_for_discount(customer):
    return is_new_customer(customer) and len(customer.get_purchase_history()) > 5

def is_eligible_for_loyalty_program(customer):
    return is_new_customer(customer) or customer.get_loyalty_level() > 3

# Benefits:
# - Single source of truth for "new customer" logic
# - Changes only need to be made in one place
# - More maintainable and less error-prone
# - Function name makes intent clear

print("✔️ Clean and modular - logic centralized in one function")
```

**📌 Key Point:** Replace temporary variables used in multiple places with functions. This creates a single source of truth, enhances modularity, and makes the code more maintainable. 🔄

## ✨ Best Practices for DRY Principle

Follow these guidelines to write DRY, maintainable code:

✅ **Identify Repetition** - Look for duplicated code blocks, expressions, or logic  
✅ **Extract Functions** - Move repeated logic into reusable functions  
✅ **Use Descriptive Names** - Make variable and function names self-documenting  
✅ **Centralize Logic** - Create single sources of truth for shared functionality  
✅ **Keep It Simple** - Don't over-engineer; refactor when you see clear duplication  
✅ **Think Ahead** - Consider future changes when deciding what to extract

## 🎓 Quick Summary

**The DRY Principle** helps create clean, maintainable code by eliminating repetition:

### Three Key Strategies:

1. **Extracting Function** → Centralize repeated logic in reusable functions
   - Reduces code bloat
   - Ensures consistency across updates
   - Changes only need to be made once

2. **Extracting Variable** → Use descriptive names for complex expressions
   - Improves code readability
   - Makes calculations self-documenting
   - Easier to understand and modify

3. **Replace Temp with Function** → Convert temporary variables to functions
   - Creates single source of truth
   - Enhances modularity
   - Eliminates duplication across methods

### Benefits of DRY:
- ✅ **Maintainability** - Changes in one place update everywhere
- ✅ **Consistency** - No risk of logic diverging across copies
- ✅ **Readability** - Clear, focused functions are easier to understand
- ✅ **Reduced Errors** - Fewer places for bugs to hide
- ✅ **Efficiency** - Less code to write, test, and debug

**Remember:** Every piece of knowledge should have a single, unambiguous representation in your codebase! 🚀

## Unit 2: The KISS Principle (Keep It Simple, Stupid)

---

## 📚 Overview

This unit covers the **KISS principle**, which champions simplicity in code design. Learn how to write straightforward, maintainable code by avoiding unnecessary complexity and focusing on clarity.

## 🔷 What is KISS?

**Definition:** "Keep It Simple, Stupid" - A principle that encourages developers to avoid unnecessary complexity and write code that is straightforward and clear.

### 💡 Core Philosophy
Simple code is better code. Complexity should only be introduced when absolutely necessary, not as a default approach.

## 🌟 Why Use KISS?

Adopting the KISS principle brings several key advantages:

✅ **Maintainer's Dream** - Simple code is easier to modify and update  
✅ **Clear Communication** - Easy to read and understand for all developers  
✅ **Testing Made Easy** - Simpler logic reduces testing complexity  
✅ **Reduced Bugs** - Less complexity means fewer places for errors to hide  
✅ **Faster Onboarding** - New team members can understand code quickly

## 🛠️ How to Apply KISS?

Key strategies for implementing the KISS principle:

✅ **Write Smaller Programs** - Keep methods and classes concise, solve one problem at a time  
✅ **Remove Unused Code** - Eliminate dead code that serves no purpose  
✅ **Focus on Readability** - Write transparent, straightforward code  
✅ **Employ Composition** - Use existing code effectively instead of rewriting  
✅ **Modular Programming** - Break applications into independent modules

## 🔴 Example #1: Temperature Conversion

**Problem:** A single function with multiple conditional branches becomes hard to maintain and extend.

### ❌ Bad Code - Complex Conditional Logic

```python
# ❌ BAD: Complex function with multiple conversion paths
def convert_temperature(temperature, from_scale, to_scale):
    if from_scale == 'F' and to_scale == 'C':
        return (temperature - 32) * 5 / 9  # Fahrenheit to Celsius
    elif from_scale == 'C' and to_scale == 'F':
        return (temperature * 9 / 5) + 32  # Celsius to Fahrenheit
    elif from_scale == 'F' and to_scale == 'K':
        return (temperature - 32) * 5 / 9 + 273.15  # Fahrenheit to Kelvin
    elif from_scale == 'K' and to_scale == 'F':
        return (temperature - 273.15) * 9 / 5 + 32  # Kelvin to Fahrenheit
    elif from_scale == 'C' and to_scale == 'K':
        return temperature + 273.15  # Celsius to Kelvin
    elif from_scale == 'K' and to_scale == 'C':
        return temperature - 273.15  # Kelvin to Celsius
    return temperature

# Problems:
# - Long chain of conditionals
# - Hard to add new temperature scales
# - Difficult to test individual conversions
# - Low readability

print("❌ Complex and hard to maintain - too many conditionals!")
```

### ✅ Good Code - Simple Functions + Dictionary Dispatch

```python
# ✅ GOOD: Simple, dedicated functions for each conversion
def fahrenheit_to_celsius(temp):
    return (temp - 32) * 5 / 9

def celsius_to_fahrenheit(temp):
    return (temp * 9 / 5) + 32

def fahrenheit_to_kelvin(temp):
    return (temp - 32) * 5 / 9 + 273.15

def kelvin_to_fahrenheit(temp):
    return (temp - 273.15) * 9 / 5 + 32

def celsius_to_kelvin(temp):
    return temp + 273.15

def kelvin_to_celsius(temp):
    return temp - 273.15

# Dictionary for function dispatching - clean and extensible
conversion_functions = {
    ('F', 'C'): fahrenheit_to_celsius,
    ('C', 'F'): celsius_to_fahrenheit,
    ('F', 'K'): fahrenheit_to_kelvin,
    ('K', 'F'): kelvin_to_fahrenheit,
    ('C', 'K'): celsius_to_kelvin,
    ('K', 'C'): kelvin_to_celsius
}

def convert_temperature(temperature, from_scale, to_scale):
    if from_scale == to_scale:
        return temperature
    convert_fn = conversion_functions.get((from_scale, to_scale))
    return convert_fn(temperature) if convert_fn else temperature

# Test it
print(f"100°F to Celsius: {convert_temperature(100, 'F', 'C'):.2f}°C")
print(f"0°C to Fahrenheit: {convert_temperature(0, 'C', 'F'):.2f}°F")
print("✔️ Simple, testable, and easy to extend!")
```

**📌 Key Point:** Breaking complex conditionals into dedicated functions with dictionary dispatch simplifies the code, makes each conversion independently testable, and makes it easy to add new temperature scales. 🌡️

## 🔴 Example #2: String Processing

**Problem:** Manual character-by-character processing with multiple buffers and conditionals adds unnecessary complexity.

### ❌ Bad Code - Over-Engineered String Processing

```python
# ❌ BAD: Over-complicated character-by-character processing
def process_string(input_string):
    # Initialize empty lists and variables to store our processed words
    words = []          # List to store final processed words
    current_word = ""   # Buffer to build each word character by character
    
    # Iterate through each character in the input string
    for i in range(len(input_string)):
        if input_string[i].isspace():  # If we encounter a space
            if current_word:  # If we have built up a word
                # Apply capitalization rules:
                # - Words longer than 2 characters: capitalize first letter
                # - Words 2 characters or less: convert to lowercase
                if len(current_word) > 2:
                    words.append(current_word[0].upper() + current_word[1:].lower())
                else:
                    words.append(current_word.lower())
                current_word = ""  # Reset the word buffer
        else:
            # Add the character to our current word buffer
            current_word += input_string[i]
    
    # Process the last word if the string doesn't end with a space
    if current_word:
        if len(current_word) > 2:
            words.append(current_word[0].upper() + current_word[1:].lower())
        else:
            words.append(current_word.lower())
    
    # Join all words with a single space, filtering out any empty strings
    return " ".join([word for word in words if word])

# Test it
result = process_string("hello WORLD to ALL")
print(f"Result: {result}")
print("❌ Too complex - hard to understand what's happening!")
```

### ✅ Good Code - Leverage Built-in Methods

```python
# ✅ GOOD: Simple and clear using Python's built-in methods
def process_string(input_string):
    # Split the input string into words
    words = input_string.split()
    
    # Process each word using a list comprehension
    processed_words = [
        word.capitalize() if len(word) > 2 else word.lower()
        for word in words
    ]
    
    # Join all words with a single space
    return " ".join(processed_words)

# Test it
result = process_string("hello WORLD to ALL")
print(f"Result: {result}")

# Benefits:
# - Leverages Python's built-in string methods
# - Clear and concise list comprehension
# - Easy to understand and modify
# - Less code = fewer bugs
# - Much more Pythonic!

print("✔️ Simple, clear, and maintainable!")
```

**📌 Key Point:** Use built-in methods and language features instead of reinventing the wheel. Python's string methods and list comprehensions provide cleaner, more readable solutions than manual character processing. 🐍

## ✨ Best Practices for KISS Principle

Follow these guidelines to keep your code simple:

✅ **Use Built-in Features** - Leverage language capabilities instead of custom solutions  
✅ **Avoid Premature Optimization** - Don't add complexity for imaginary performance gains

## 🎓 Quick Summary

**The KISS Principle** promotes simplicity and clarity in code:

### Key Concepts:

   - Fewer bugs and easier debugging

   - Replace long conditional chains with function lookups

3. **Leverage Built-in Features** → Use language capabilities effectively
   - String methods (`.split()`, `.capitalize()`, `.lower()`)

### Benefits of KISS:

# 🎯 Applying Clean Code Principles in Python
## Unit 3: The Law of Demeter

---

## 📚 Overview

This unit covers the **Law of Demeter**, a fundamental design guideline in object-oriented programming. Learn how to reduce coupling between objects by limiting their knowledge about each other's internal structures, leading to more maintainable and modular code.

## 🔷 What is the Law of Demeter?

**Definition:** Also known as the "Principle of Least Knowledge," it states that an object should only communicate with its immediate collaborators, not with the entire system.

### 💡 Core Philosophy
Each object should have limited knowledge about other objects. An object should only talk to its friends, not to strangers.

### The Five Rules

A method **X** of class **C** should only call methods of:

1. **Class C itself** - Own methods
2. **Objects created by X** - Objects it instantiates
3. **Objects passed as arguments to X** - Direct parameters
4. **Objects held in instance variables of C** - Own attributes
5. **Static fields** - Class variables

By controlling interactions between objects, you create more organized, loosely-coupled code that's easier to maintain and scale. 🚀

## 🔴 Rule #1: Call Own Class Methods

**Principle:** A method should be able to call other methods within its own class.

### ✅ Example - Car Starting Sequence

```python
# ✅ GOOD: Method interacts only with its own class methods
class Car:
    def start(self):
        """Start method calls other methods within the Car class"""
        self.check_fuel()  # Own method
        self.ignite()      # Own method
    
    def check_fuel(self):
        print("Checking fuel level...")
    
    def ignite(self):
        print("Igniting the engine...")

# Test it
car = Car()
car.start()
print("✔️ Clear boundaries - the 'start' method only talks to its own methods!")
```

**📌 Key Point:** Methods should freely interact with their own class methods, maintaining clear internal boundaries while encapsulating the class's behavior. 🚗

## 🔴 Rule #2: Interact with Objects You Create

**Principle:** A method can call methods on objects it instantiates.

### ✅ Example - Library Borrowing Books

```python
# ✅ GOOD: Method interacts with objects it creates
class Library:
    def borrow_book(self, title):
        """Library creates a Book instance and interacts with it"""
        book = Book(title)  # Object created by this method
        book.issue()        # Can call methods on the created object
        return book

class Book:
    def __init__(self, title):
        self.title = title
    
    def issue(self):
        print(f"Book issued: {self.title}")

# Test it
library = Library()
my_book = library.borrow_book("Clean Code")
print("✔️ Library only interacts with the Book it created!")
```

**📌 Key Point:** When a method creates an object, it's allowed to interact with it. This maintains control over the object lifecycle and keeps dependencies clear. 📚

## 🔴 Rule #3: Interact with Objects Passed as Arguments

**Principle:** A method can call methods on objects passed to it as parameters.

### ✅ Example - Printer Handling Documents

```python
# ✅ GOOD: Method interacts with objects passed as parameters
class Printer:
    def print(self, document):
        """Printer can call methods on the document parameter"""
        document.send_to_printer()  # Interacts with passed argument

class Document:
    def send_to_printer(self):
        print("Document is being printed...")

# Test it
printer = Printer()
doc = Document()
printer.print(doc)
print("✔️ Printer only interacts with the document passed to it!")
```

**📌 Key Point:** Methods can interact directly with their parameters. This enables dependency injection and keeps communication limited to explicitly passed objects. 🖨️

## 🔴 Rule #4: Interact with Instance Variables

**Principle:** A method can call methods on objects held in instance variables of its class.

**⚠️ Important:** While accessing instance variables is allowed, avoid chaining multiple method calls or accessing deep internals to prevent tight coupling.

### ✅ Example - House Locking Its Door

```python
# ✅ GOOD: Method interacts with instance variables
class House:
    def __init__(self):
        self.door = Door()  # Instance variable
    
    def lock_house(self):
        """House can call methods on its instance variable (door)"""
        self.door.close()  # OK: One level of interaction

class Door:
    def close(self):
        print("Door is closed.")

# Test it
house = House()
house.lock_house()
print("✔️ House interacts with its own door instance!")
print("⚠️ Avoid: house.door.lock.mechanism.engage() - too deep!")
```

**📌 Key Point:** Methods can interact with instance variables, but keep it to one level. Avoid deep chaining (`object.a.b.c.method()`) as it creates tight coupling. 🏠

## 🔴 Rule #5: Interact with Static Fields (Class Variables)

**Principle:** Methods can access static fields (class variables).

**⚠️ Caution:** Use static fields carefully as they can lead to shared state issues in larger applications.

### ✅ Example - Temperature Converter

```python
# ✅ GOOD: Method accessing class variable (static field)
class TemperatureConverter:
    conversion_factor = 9.0 / 5.0  # Class variable (static field)
    
    @classmethod
    def celsius_to_fahrenheit(cls, celsius):
        """Accesses the class variable for conversion"""
        return int((celsius * cls.conversion_factor) + 32)

# Test it
temp_f = TemperatureConverter.celsius_to_fahrenheit(100)
print(f"100°C = {temp_f}°F")
print("✔️ Method uses static field for the conversion factor!")
```

**📌 Key Point:** Static fields are acceptable for constants and shared data, but use them cautiously to avoid shared state problems in complex applications. 🌡️

## 🚫 Violation Example: Tight Coupling

**Problem:** `Person` directly accesses multiple internal methods of `Address`, creating tight coupling and fragile code.

### ❌ Bad Code - Violating Law of Demeter

```python
# ❌ BAD: Person knows too much about Address internals
class Person:
    def __init__(self, address):
        self.address = address
    
    def get_address_details(self):
        # Violation: Accessing multiple internal methods of Address
        return ("Address: " + 
                self.address.get_first_name() + " " +      # Too much knowledge
                self.address.get_last_name() +             # About Address
                ", " + self.address.get_street() +         # Internal structure
                ", " + self.address.get_city() + 
                ", " + self.address.get_country() + 
                ", ZipCode: " + self.address.get_zip_code())

class Address:
    def __init__(self, first_name, last_name, street, city, country, zip_code):
        self.first_name = first_name
        self.last_name = last_name
        self.street = street
        self.city = city
        self.country = country
        self.zip_code = zip_code
    
    def get_first_name(self):
        return self.first_name
    
    def get_last_name(self):
        return self.last_name
    
    def get_street(self):
        return self.street
    
    def get_city(self):
        return self.city
    
    def get_country(self):
        return self.country
    
    def get_zip_code(self):
        return self.zip_code

# Problems:
# - Person relies on Address internal structure
# - Changes to Address break Person
# - Tight coupling makes code fragile
# - Violates encapsulation

print("❌ Tight coupling - Person knows too much about Address!")
```

### ✅ Good Code - Refactored with Proper Encapsulation

```python
# ✅ GOOD: Address encapsulates its own formatting logic
class Person:
    def __init__(self, address):
        self.address = address
    
    def get_address_details(self):
        # Simple: Just ask Address for its formatted string
        return self.address.get_address_line()  # Single method call!

class Address:
    def __init__(self, first_name, last_name, street, city, country, zip_code):
        self.first_name = first_name
        self.last_name = last_name
        self.street = street
        self.city = city
        self.country = country
        self.zip_code = zip_code
    
    def get_address_line(self):
        """Address knows how to format itself"""
        return (self.first_name + " " + self.last_name +
                ", " + self.street +
                ", " + self.city + 
                ", " + self.country +
                ", ZipCode: " + self.zip_code)
    
    # Internal methods for Address's own use
    def get_first_name(self):
        return self.first_name
    
    def get_last_name(self):
        return self.last_name
    
    def get_street(self):
        return self.street
    
    def get_city(self):
        return self.city
    
    def get_country(self):
        return self.country
    
    def get_zip_code(self):
        return self.zip_code

# Test it
address = Address("John", "Doe", "123 Main St", "Boston", "USA", "02101")
person = Person(address)
print(person.get_address_details())

# Benefits:
# - Person doesn't know Address internal structure
# - Changes to Address don't affect Person
# - Loose coupling makes code maintainable
# - Each class has clear responsibilities

print("✔️ Proper encapsulation - Person only talks to Address, not its internals!")
```

**📌 Key Point:** Encapsulate complex logic within the appropriate class. By having `Address` provide a single `get_address_line()` method, `Person` no longer needs to know about Address internals, reducing coupling and improving maintainability. 🎯

## ✨ Best Practices for Law of Demeter

Follow these guidelines to reduce coupling and improve code modularity:

✅ **Talk to Friends, Not Strangers** - Only communicate with immediate collaborators  
✅ **Avoid Method Chaining** - Limit deep navigation like `object.a.b.c.method()`  
✅ **Encapsulate Internal Logic** - Each class should format/manage its own data  
✅ **Single Level of Interaction** - Call methods on direct dependencies only  
✅ **Reduce Knowledge** - Objects shouldn't know about the internal structure of other objects  
✅ **Use Delegation** - Let objects handle their own responsibilities  
✅ **Minimize Dependencies** - Less coupling means easier maintenance and testing

## 🎓 Quick Summary

**The Law of Demeter** (Principle of Least Knowledge) ensures loose coupling by limiting object interactions:

### The Five Rules:

1. **Call Own Methods** → Methods can call other methods in the same class
   - Maintains clear internal boundaries
   - Encapsulates class behavior

2. **Interact with Created Objects** → Methods can call methods on objects they instantiate
   - Controlled object lifecycle
   - Clear dependency management

3. **Interact with Parameters** → Methods can call methods on objects passed as arguments
   - Enables dependency injection
   - Explicit communication channels

4. **Interact with Instance Variables** → Methods can call methods on their class attributes
   - Keep it to one level (avoid chaining)
   - Prevents tight coupling

5. **Access Static Fields** → Methods can use class variables
   - Good for constants
   - Use cautiously to avoid shared state issues

### Key Violations to Avoid:
- ❌ **Method Chaining** - `object.a.b.c.method()` creates tight coupling
- ❌ **Accessing Internals** - Calling multiple getters on other objects
- ❌ **Deep Navigation** - Reaching through objects to their dependencies

### Benefits of Law of Demeter:
- ✅ **Loose Coupling** - Objects are independent and modular
- ✅ **Easier Maintenance** - Changes don't ripple through the codebase
