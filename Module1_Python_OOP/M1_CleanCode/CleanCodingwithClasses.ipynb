# Clean Coding with Classes 🏗️

## Unit 1: Single Responsibility Principle (SRP)

### What is SRP? 🎯
The Single Responsibility Principle is one of the most important principles in clean code, stating that:
> "A class should have only one reason to change"

This means each class should have:
- Only one job
- One clear purpose
- One responsibility

### Why is SRP Important? 🤔
Following SRP leads to:
- ✅ More maintainable code
- ✅ Easier testing
- ✅ Better code organization
- ✅ Reduced bugs
- ✅ Simpler updates

````python
# Example of a class that violates SRP
class Report:
    def generate_report(self):
        # Handles report generation
        return "Sales Report Data"
    
    def print_report(self, report_content):
        # Handles printing
        print(report_content)
    
    def save_to_file(self, report_content, file_path):
        # Handles file operations
        print(f"Saving report to {file_path}...")
    
    def send_by_email(self, email, report_content):
        # Handles email operations
        print(f"Sending email to {email}")

# Usage of the non-SRP class
report = Report()
content = report.generate_report()
report.print_report(content)
report.save_to_file(content, "report.txt")
report.send_by_email("user@example.com", content)
````

````python
# Refactored version following SRP
class ReportGenerator:
    def generate(self):
        """Handles only report generation"""
        return "Sales Report Data"

class ReportPrinter:
    def print_report(self, report_content):
        """Handles only printing"""
        print(report_content)

class ReportSaver:
    def save_to_file(self, report_content, file_path):
        """Handles only file operations"""
        print(f"Saving report to {file_path}...")

class EmailSender:
    def send_by_email(self, email, report_content):
        """Handles only email operations"""
        print(f"Sending email to {email}")

# Usage of SRP-compliant classes
generator = ReportGenerator()
printer = ReportPrinter()
saver = ReportSaver()
emailer = EmailSender()

# Each class has a single responsibility
content = generator.generate()
printer.print_report(content)
saver.save_to_file(content, "report.txt")
emailer.send_by_email("user@example.com", content)
````

# Unit 2: Encapsulation in Python 🛡️

## What is Encapsulation? 🎯
Encapsulation is a fundamental concept in object-oriented programming that:
- Bundles data and methods that operate on that data into a single unit
- Controls access to an object's attributes and methods
- Protects internal state of objects

### Benefits of Encapsulation:
- ✅ Simplified maintenance
- ✅ Prevention of data misuse
- ✅ Enhanced security
- ✅ Better control over data access
- ✅ Reduced debugging complexity

````python
# Example of Poor Encapsulation
class Book:
    def __init__(self):
        self.title = None
        self.author = None
        self.price = None

# Usage showing problems with poor encapsulation
book = Book()
book.title = "Clean Code"
book.author = "Robert C. Martin"
book.price = -10.0  # Problem: Can set invalid price!

print(f"Book price: {book.price}")  # Prints negative price without any validation
````

````python
# Example of Good Encapsulation using Properties
class Book:
    def __init__(self, title, author, price):
        self._title = title      # Protected attribute (convention)
        self._author = author    # Protected attribute (convention)
        self._price = price      # Protected attribute (convention)
    
    @property
    def title(self):
        return self._title
    
    @property
    def author(self):
        return self._author
    
    @property
    def price(self):
        return self._price
    
    @price.setter
    def price(self, value):
        if value >= 0:
            self._price = value
        else:
            raise ValueError("Price cannot be negative")

# Usage of well-encapsulated class
try:
    # Create a book with valid data
    book = Book("Clean Code", "Robert C. Martin", 29.99)
    print(f"Current price: ${book.price}")
    
    # Try to set an invalid price
    book.price = -10.0  # This will raise an error
except ValueError as e:
    print(f"Error: {e}")  # Will print: "Error: Price cannot be negative"
````

### Quick Guide to Python Encapsulation Features 🔍

1. **Underscore Prefix `_`**
   - `self._name`: Indicates a "protected" attribute
   - It's a convention to say "this is for internal use"
   - Other developers shouldn't access these directly
   - Example: `self._price = price`

2. **@property Decorator**
   - Turns a method into a getter
   - Allows you to access method as if it were an attribute
   - Instead of `book.get_price()`, you use `book.price`
   ```python
   @property
   def price(self):
       return self._price
   ```

3. **@attribute.setter Decorator**
   - Defines how an attribute should be set
   - Perfect for adding validation
   - Protects data integrity
   ```python
   @price.setter
   def price(self, value):
       if value >= 0:
           self._price = value
   ```

These features help you write cleaner, safer code without sacrificing Python's simplicity! 🐍

# Unit 3: Constructors and Object Initialization 🏗️

## What are Clean Constructors? 🎯
The `__init__` method in Python is crucial for initializing objects correctly and maintaining clean code principles.

### Key Concepts:
- Constructors initialize object state
- They ensure object validity from creation
- They establish object dependencies

### Benefits of Clean Constructors:
- ✅ Clear object initialization
- ✅ Predictable object state
- ✅ Better error handling
- ✅ Easier testing
- ✅ Improved maintainability

````python
# Example of Poor Constructor Design
class UserProfile:
    def __init__(self, data_string):
        # Complex parsing in constructor
        data = data_string.split(',')
        self.name = data[0]
        self.email = data[1]
        self.age = int(data[2])
        self.address = data[3]
        self.validate_complex_data()  # Complex validation
        self.setup_additional_data()  # Additional setup
    
    def validate_complex_data(self):
        # Complex validation logic in constructor
        pass
    
    def setup_additional_data(self):
        # More setup logic in constructor
        pass

# Usage with complex, error-prone initialization
data = "John Doe,john@example.com,30,123 Main St"
user = UserProfile(data)  # Multiple things could go wrong here
````

````python
# Example of Clean Constructor Design
class UserProfile:
    def __init__(self, name, email, age, address):
        """Simple constructor with clear parameters"""
        self.name = name
        self.email = email
        self.age = age
        self.address = address
    
    @classmethod
    def from_string(cls, data_string):
        """Factory method for complex initialization"""
        data = data_string.split(',')
        return cls(
            name=data[0],
            email=data[1],
            age=int(data[2]),
            address=data[3]
        )

# Clean usage with explicit constructor
user1 = UserProfile("John Doe", "john@example.com", 30, "123 Main St")

# Alternative creation using factory method for complex cases
data = "Jane Doe,jane@example.com,25,456 Oak St"
user2 = UserProfile.from_string(data)
````

### Quick Guide to Constructor Features in Python 🔍

1. **@classmethod Decorator**
   - Creates alternative constructors
   - Uses `cls` instead of `self`
   - Allows different ways to create objects
   - Perfect for complex initialization logic
   ```python
   @classmethod
   def from_string(cls, data_string):
       # Process data_string
       return cls(processed_data)
   ```

2. **Factory Methods**
   - Are class methods that create objects
   - Keep `__init__` simple and clean
   - Handle complex creation logic
   - Example:
   ```python
   @classmethod
   def from_dictionary(cls, data_dict):
       return cls(
           name=data_dict['name'],
           age=data_dict['age']
       )
   ```

3. **When to Use Class Methods as Constructors:**
   - When you need different ways to create objects
   - To handle complex initialization logic
   - To convert from different data formats
   - To keep `__init__` simple and focused

Remember: Class methods help maintain clean constructors by moving complex initialization logic out of `__init__`! 🎯

````python
# Example of Poor Constructor Design
class UserProfile:
    def __init__(self, data_string):
        # Complex parsing in constructor
        data = data_string.split(',')
        self.name = data[0]
        self.email = data[1]
        self.age = int(data[2])
        self.address = data[3]

# Usage of poor design
data = "John Doe,john@example.com,30,123 Main St"
user = UserProfile(data)
print(f"Name: {user.name}")
print(f"Age: {user.age}")  # Could raise error if data format is wrong!
````

````python
# Example of Clean Constructor Design
class UserProfile:
    def __init__(self, name, email, age, address):
        """Simple constructor with clear parameters"""
        self._validate_age(age)
        self.name = name
        self.email = email
        self.age = age
        self.address = address
    
    @classmethod
    def from_string(cls, data_string):
        """Factory method for creating from string"""
        data = data_string.split(',')
        return cls(
            name=data[0],
            email=data[1],
            age=int(data[2]),
            address=data[3]
        )
    
    def _validate_age(self, age):
        """Validate age is reasonable"""
        if not isinstance(age, (int, float)) or age < 0:
            raise ValueError("Age must be a positive number")

# Clean usage with explicit parameters
user1 = UserProfile("John Doe", "john@example.com", 30, "123 Main St")

# Alternative creation using factory method
data = "Jane Doe,jane@example.com,25,456 Oak St"
user2 = UserProfile.from_string(data)

print(f"User 1: {user1.name}, {user1.age} years old")
print(f"User 2: {user2.name}, {user2.age} years old")
````

````python
# Example of Builder Pattern for Complex Objects
class Pizza:
    def __init__(self, builder):
        """Constructor only receives a builder"""
        self.size = builder.size
        self.crust = builder.crust
        self.toppings = builder.toppings.copy()
    
    class Builder:
        def __init__(self, size):
            """Initialize with required parameter"""
            self.size = size
            self.crust = "regular"  # default value
            self.toppings = []
        
        def set_crust(self, crust):
            """Optional: set crust type"""
            self.crust = crust
            return self
        
        def add_topping(self, topping):
            """Optional: add a topping"""
            self.toppings.append(topping)
            return self
        
        def build(self):
            """Create the final object"""
            return Pizza(self)

# Using the builder pattern
pizza = (Pizza.Builder("large")
         .set_crust("thin")
         .add_topping("cheese")
         .add_topping("mushrooms")
         .add_topping("pepperoni")
         .build())

print(f"Pizza size: {pizza.size}")
print(f"Pizza crust: {pizza.crust}")
print(f"Pizza toppings: {', '.join(pizza.toppings)}")
````
