# Clean Coding with Classes 🏗️

## Unit 1: Single Responsibility Principle (SRP)

### What is SRP? 🎯
The Single Responsibility Principle is one of the most important principles in clean code, stating that:
> "A class should have only one reason to change"

This means each class should have:
- Only one job
- One clear purpose
- One responsibility

### Why is SRP Important? 🤔
Following SRP leads to:
- ✅ More maintainable code
- ✅ Easier testing
- ✅ Better code organization
- ✅ Reduced bugs
- ✅ Simpler updates

````python
# Example of a class that violates SRP
class Report:
    def generate_report(self):
        # Handles report generation
        return "Sales Report Data"
    
    def print_report(self, report_content):
        # Handles printing
        print(report_content)
    
    def save_to_file(self, report_content, file_path):
        # Handles file operations
        print(f"Saving report to {file_path}...")
    
    def send_by_email(self, email, report_content):
        # Handles email operations
        print(f"Sending email to {email}")

# Usage of the non-SRP class
report = Report()
content = report.generate_report()
report.print_report(content)
report.save_to_file(content, "report.txt")
report.send_by_email("user@example.com", content)
````

````python
# Refactored version following SRP
class ReportGenerator:
    def generate(self):
        """Handles only report generation"""
        return "Sales Report Data"

class ReportPrinter:
    def print_report(self, report_content):
        """Handles only printing"""
        print(report_content)

class ReportSaver:
    def save_to_file(self, report_content, file_path):
        """Handles only file operations"""
        print(f"Saving report to {file_path}...")

class EmailSender:
    def send_by_email(self, email, report_content):
        """Handles only email operations"""
        print(f"Sending email to {email}")

# Usage of SRP-compliant classes
generator = ReportGenerator()
printer = ReportPrinter()
saver = ReportSaver()
emailer = EmailSender()

# Each class has a single responsibility
content = generator.generate()
printer.print_report(content)
saver.save_to_file(content, "report.txt")
emailer.send_by_email("user@example.com", content)
````

# Unit 2: Encapsulation in Python 🛡️

## What is Encapsulation? 🎯
Encapsulation is a fundamental concept in object-oriented programming that:
- Bundles data and methods that operate on that data into a single unit
- Controls access to an object's attributes and methods
- Protects internal state of objects

### Benefits of Encapsulation:
- ✅ Simplified maintenance
- ✅ Prevention of data misuse
- ✅ Enhanced security
- ✅ Better control over data access
- ✅ Reduced debugging complexity

````python
# Example of Poor Encapsulation
class Book:
    def __init__(self):
        self.title = None
        self.author = None
        self.price = None

# Usage showing problems with poor encapsulation
book = Book()
book.title = "Clean Code"
book.author = "Robert C. Martin"
book.price = -10.0  # Problem: Can set invalid price!

print(f"Book price: {book.price}")  # Prints negative price without any validation
````

````python
# Example of Good Encapsulation using Properties
class Book:
    def __init__(self, title, author, price):
        self._title = title      # Protected attribute (convention)
        self._author = author    # Protected attribute (convention)
        self._price = price      # Protected attribute (convention)
    
    @property
    def title(self):
        return self._title
    
    @property
    def author(self):
        return self._author
    
    @property
    def price(self):
        return self._price
    
    @price.setter
    def price(self, value):
        if value >= 0:
            self._price = value
        else:
            raise ValueError("Price cannot be negative")

# Usage of well-encapsulated class
try:
    # Create a book with valid data
    book = Book("Clean Code", "Robert C. Martin", 29.99)
    print(f"Current price: ${book.price}")
    
    # Try to set an invalid price
    book.price = -10.0  # This will raise an error
except ValueError as e:
    print(f"Error: {e}")  # Will print: "Error: Price cannot be negative"
````

### Quick Guide to Python Encapsulation Features 🔍

1. **Underscore Prefix `_`**
   - `self._name`: Indicates a "protected" attribute
   - It's a convention to say "this is for internal use"
   - Other developers shouldn't access these directly
   - Example: `self._price = price`

2. **@property Decorator**
   - Turns a method into a getter
   - Allows you to access method as if it were an attribute
   - Instead of `book.get_price()`, you use `book.price`
   ```python
   @property
   def price(self):
       return self._price
   ```

3. **@attribute.setter Decorator**
   - Defines how an attribute should be set
   - Perfect for adding validation
   - Protects data integrity
   ```python
   @price.setter
   def price(self, value):
       if value >= 0:
           self._price = value
   ```

These features help you write cleaner, safer code without sacrificing Python's simplicity! 🐍
