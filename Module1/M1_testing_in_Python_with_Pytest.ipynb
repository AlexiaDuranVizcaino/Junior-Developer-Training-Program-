# Introduction 

### What is Pytest?
- **Testing framework for Python** – makes it simple to write and run tests.  
- **Auto-discovery of tests** – pytest automatically detects tests based on their file and function names, so no extra configuration is needed.  
- **Rich assertion introspection** – when a test fails, pytest provides clear and detailed error messages, making it easy to understand what went wrong.  
- **Support for parameterized and fixture-based testing** – allows running the same test with multiple inputs and reusing setups to keep tests clean and maintainable.  


### Why Pytest?
- **Simplified syntax** – easy to learn and requires less boilerplate code.  
- **Rich assertion introspection** – provides detailed error messages when a test fails.  
- **Powerful fixture system** – allows reusable setups and clean test environments.

- **Extensibility** – supports many plugins and custom extensions.

# 2. Class based Tests

We use a class-based structure in pytest to organize related tests for the Circle class. The TestCircle class contains setup and teardown methods and two test cases: test_area and test_perimeter.

### 1. Setup Method 
The setup_method runs before each test. Here, it creates a Circle instance with a radius of 10 and prints a message showing which test is being set up. This ensures each test starts with a fresh object.


```python
def setup_method(self, method):
    print(f"Setting up {method}")
    self.circle = shapes.Circle(10)
```

### 2.TearDown Method 
The teardown_method runs after each test. It deletes the Circle instance and prints a message indicating the test being torn down. This cleans up resources and prevents side effects between tests.

```python
def teardown_method(self, method):
    print(f"Tearing down {method}")
    del self.circle
```

### 3. Test Methods 
- **test_area**: Verifies that the `area()` method of the `Circle` class returns the correct value using the formula \( pi r^2 \).  
- **test_perimeter**: Checks that the `perimeter()` method returns the correct value using the formula \( 2 pi r \).


```python
def test_area(self):
    assert self.circle.area() == math.pi * self.circle.radius ** 2

def test_perimeter(self):
    assert self.circle.perimeter() == 2 * math.pi * self.circle.radius
```

### Output 
When running pytest with -s, messages from setup and teardown show which test is being prepared and cleaned up. Both tests pass successfully.

- **setup_method** → initializes the `Circle` object before each test.  
- **teardown_method** → deletes the `Circle` object after each test to clean up resources.  
- **test_area** → checks that the `area()` method returns the correct value (\( \pi r^2 \)).  
- **test_perimeter** → checks that the `perimeter()` method returns the correct value (\( 2 \pi r \)).


# 3. Fixtures
**Fixtures** in pytest are reusable setups that provide test data or objects. They are passed as parameters to test functions and help avoid repeating code. You can define them in a separate script and use them across your tests.

- Fixtures can be reused in multiple tests.  
- They can also be used in class-based tests.  
- Using fixtures ensures tests are clean, maintainable, and consistent.

```python
import pytest
@pytest.fixture

def my_rectangle():
    return shapes.Rectangle(10, 20)

def test_area(my_rectangle):
    assert my_rectangle.area() == 10 * 20

def test_not_same_area_rectangle(self, my_rectangle):
    assert self.circle.area() != my_rectangle.area()
```

# 4. Marks and Prametrization

### Marks in Pytest

Marks in pytest are special labels that you can add to your test functions to control or organize how they are executed.  
They help you **skip**, **expect failures**, or **categorize** tests.

**Common examples:**

- `@pytest.mark.skip(reason="...")` → skips the test temporarily (useful when a feature is not ready or broken).  
- `@pytest.mark.xfail(reason="...")` → marks a test that is expected to fail (e.g., known bug).  
- `@pytest.mark.slow` → labels a test as slow, so you can run or exclude it with specific options.


### Parametrization in Pytest

`@pytest.mark.parametrize` allows you to run the **same test function multiple times** with different sets of input values and expected results.  

Each tuple inside the list represents one test case, and pytest automatically iterates through all of them. This makes your tests **more efficient**, **readable**, and **less repetitive**, since you don’t need to write separate functions for each scenario.

For example, in our code, the test is executed several times with different `side_length` and expected results for both **area** and **perimeter** calculations.  
Each successful test case is represented by a dot (`.`) in the pytest output.


```python
import pytest 

@pytest.mark.parametrize("side_length, expected_area", [(5, 25), (4, 16), (9, 81)])
def test_multiple_square_areas(side_length, expected_area):
    assert shapes.Square(side_length).area() == expected_area

@pytest.mark.parametrize("side_length, expected_perimeter", [(3,12), (4, 16), (5, 20)])
def test_multiple_perimeters(side_length, expected_perimeter):
    assert shapes.Square(side_length).perimeter() == expected_perimeter
```

# 5. Mocking 

Mocking allows us to replace external functions or services with fake objects during tests. This is useful to test our code without relying on real databases or APIs, which might be slow or fail. By using mocks, we can simulate different scenarios such as successful operations, errors, or edge cases, and ensure that our code behaves correctly in each situation.


The examples included demonstrate:

- **Replacing `get_user_from_db` with a mock that always returns `"Mocked Alice"`**  
  This allows testing code that depends on the database without accessing the real one. It ensures tests are consistent and not affected by changes in the database.

- **Replacing `requests.get` with a mock that simulates API responses**  
  - **Success case (`status_code = 200` with JSON data):** tests how the function handles correct API responses.  
  - **Error case (`status_code = 400`):** tests exception handling using `pytest.raises(requests.HTTPError)` without calling the real API.

Using mocks in these ways allows us to test different scenarios (success, error, edge cases) quickly and reliably.

```python
import requests

database = {
    1: "Alice",
    2: "Bob",
    3: "Charlie"
}

def get_user_from_db(user_id):
    return database.get(user_id)

def get_users():
    response = requests.get("https://jsonplaceholder.typicode.com/users")
    if response.status_code == 200:
        return response.json()
    
    raise requests.HTTPError
```

```python
import pytest 
from Source import service as service
import unittest.mock as mock 
import requests


@mock.patch("Source.service.get_user_from_db")
def test_get_user_from_db(mock_get_user_from_db):
    mock_get_user_from_db.return_value = "Mocked Alice"
    user_name = service.get_user_from_db(1)

    assert user_name == "Mocked Alice"

@mock.patch("requests.get")
def test_get_users(mock_get):
    mock_response = mock.Mock()
    mock_response.status_code = 200
    mock_response.json.return_value = {"id": 1, "name": "John Doe"}
    mock_get.return_value = mock_response
    data = service.get_users()
    assert data ==  {"id": 1, "name": "John Doe"}

@mock.patch("requests.get")
def test_get_users_errors(mock_get):
    mock_response = mock.Mock()
    mock_response.status_code = 400
    mock_get.return_value = mock_response
    with pytest.raises(requests.HTTPError):
        service.get_users()
```
