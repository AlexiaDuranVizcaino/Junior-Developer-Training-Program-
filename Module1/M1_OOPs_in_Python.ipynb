# Object Oriented Programming 

A class is like a template for creating objects.
For example: a Phone class could have properties like color and size, and behaviors like make_calls() or play_games().
In Python, defining a class is like creating a new data type (e.g., Mobile).
The attributes represent the properties, and the methods represent the behaviors.

Objects are specific instances of a class.
For example, if Phone is a class, then an iPhone or a Samsung would be individual objects created from that class — each with its own specific values for the attributes (like color, size, etc.).

```python
class Phone:

    def playing_game(self):
        print("Im playing a game")

    def make_call(self):
        print("Making a call")
```

```python
p1 = Phone()
p1.playing_game()
```

```python
p1.make_call()
```

```python
class Phone:

    def playing_game(self):
        print("Im playing a game")

    def make_call(self):
        print("Making a call")

    def assign_color(self, color):
        self.color = color
    
    def assign_cost(self, cost):
        self.cost = cost 

    def show_color(self):
        return self.color 
    
    def show_cost(self):
        return self.cost 
```

```python
p2 = Phone()
```

```python
p2.assign_color("Red")
p2.assign_cost(100)
```

```python
p2.show_color()
```

```python
p2.show_cost()
```

## Instantiating Object 
The `__init__` method acts as the constructor of a class.  
It is automatically called when a new object is created and is used to initialize the object's attributes.

```python
class Employee:
    def __init__(self, name, age, salary, gender):

        self.name = name 
        self.age = age 
        self.salary = salary 
        self.gender = gender 

    def employee_details(self):
        print("Name of the employee is::",self.name)
        print("Age of the employee is::",self.age)
        print("Salary of the employee is::",self.salary)
        print("Gender of the employee is::",self.gender)
```

```python
e1 = Employee("Alexia", 24, 150000, "Female")
```

```python
e1.employee_details()
```

## Inheritance
With inheritance one class can derive the properties of another class.

```python
class Vehicle:

    def __init__(self, mileage, cost):
        self.mileage = mileage 
        self.cost = cost

    def show_vehicle_details(self):
        print("Mileage of the vehicle is :",self.mileage)
        print("Cost of the vehicle is :",self.cost)
```

```python
class Car(Vehicle):

    def show_car_details(self):
        print("I am a car")
```

```python
c1 = Car(500,30000)
```

```python
c1.show_vehicle_details()
```

```python
c1.show_car_details()
```

# Overriding the `__init__` Method

When overriding the `__init__` method in a child class, you can use `super().__init__()` to call the parent class’s constructor.  
This allows you to pass attributes from the parent class to the child class.

# Types of Inheritance

- **Single Inheritance:** A child class inherits from a single parent class.  
- **Multiple Inheritance:** A child class inherits from more than one parent class.  
- **Multilevel Inheritance:** There is a parent, child, and grandchild class in a hierarchical relationship.  
- **Hybrid Inheritance:** A combination of two or more types of inheritance.


```python
class Car(Vehicle):

    def __init__(self, mileage, cost, tyres, hp):
        super().__init__(mileage, cost)
        self.tyres = tyres
        self.hp = hp

    def show_details(self):
        print("Im a car")
        print("Number of tyres are",self.tyres)
        print("Value of horse power is", self.hp)
```

```python
c2 =  Car(20, 120000, 4, 300)
```

```python
c2.show_details()
```

```python
c2.show_vehicle_details()
```
